```python
query = "SELECT DISTINCT country FROM elect"
pd.read_sql(query, db)

# DISTINCT — SQL’da takroriy (dublikat) qiymatlarni olib tashlab,
# faqat noyob (unique) qiymatlarni chiqarish uchun ishlatiladi.
```

```python
countries = pd.read_sql(query, db)
```

```python
'Uzbekistan' in countries['country'].values
# yuqoridagi jadvalda o'zbekiston bor yo'qligini tekshiryabmiz True chiqdi yani bor ekan
```

# **19-dars **
# Between, in va like

```python
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2005 AND 2015"
pd.read_sql(query, db)
# BETWEEN bilan yili 2005 va 2015 yillar oralig'ini olyabmiz
```

```python
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year IN (1990,1995,2000,2005,2010,1985)"
pd.read_sql(query, db)
# IN bilan faqat biz yozgan yillar malumotini olishimiz mumkin
```

```python
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year IN (2027,2036)"
pd.read_sql(query, db)
# Agar mavjud bo'lmagan yillar kiritsak hech qanday qiymat qaytarmaydi chunki bu yillar jadvalda mavjud emas
```

```python
# LIKE
# %var  =>  Anvar, Sarvar      var dan oldin boshqa harfli so'zlar ham bor degani
# _ar   =>  car, bar, tar      yani pastki chiziqcha bitta harf bor degani o'rniga
# __ar  =>  star               bunda 2 ta pastki chiziqcha qildik 2 ta harf bor degani
# %ol%  =>  bola, hola, bolalar    ikki tarafida ham so'zlar mavjud degani

query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019,2020,2021)"
pd.read_sql(query, db)
# country si oxiri %stan bilan tugagan davlatlarnichiqardi
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year=2020"
pd.read_sql(query, db)
# code ustunidagi __A dan oldin 2 ta xarf bo'lgan qatorlarni  chiqardi
```
