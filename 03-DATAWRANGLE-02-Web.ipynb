### Web fayllar bilan ishlash
#### `read_html()` - web sahifadagi barcha jadvallarni o'qib, `DataFrame1lardan iborat ro'yxat ko'rinishida qaytaradi.

```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Tashkent"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)
data = pd.read_html(response.text)
```

`data` tarkibida sahifadagi barcha jadvallar alohida `DataFrame` ko'rinishida saqlangan. Jadvallar sonini ko'rish uchun `len()` funksiyasiga murojaat qilamiz.

```python
len(data)
```

### JSON format

JSON web orqali ma'lumotlarni uzatishda ishlatiluvchi eng mashhur formatlardan. JSON haqida Dasturlash Asoslari darsimizda batafsil ma'lumot berilgan: https://python.sariq.dev/files-exceptions/34-json

```python
url = "http://api.nobelprize.org/v1/laureate.json"
r = requests.get(url)
data = r.json()
df_laureates = pd.DataFrame(data['laureates'])
display(df_laureates)
```
