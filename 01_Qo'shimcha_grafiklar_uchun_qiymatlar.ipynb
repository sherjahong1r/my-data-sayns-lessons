```python
x = np.arange(0,  4*np.pi, 0.1) # x nuqtasi 0 dan 4*pi gacha 0.1 oraliqda
sin = np.sin(x) # yuqoridagi x qiymatlarini sin qiymatida aniqladik sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6)) # 12 ga 6 hajmi
# sin(x) grafigi va ko'rinishini belgilovchi qiymatlar
plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         color='#8E44AD',
         alpha=0.7,        # bu shafoflik darajasi 0 dan 1 gacha agar 0 qilsak ko'rinmas edi.
         marker='h')       # nuqtalarini oltiburchak shaklida belgilash uchun

# cos(x) grafigi va ko'rinishini belgilovchi qiymatlar
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')      # pastga qaragan uchburchak bo'ladi

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x - o'qi nomi
plt.xlabel("pi o'qi")
# y - o'qi nomi
plt.ylabel("cos and sin o'qlar")
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x) funksiya','cos(x) funksiya'], loc=0, fontsize='x-large')
# legaend bilan sin va cos qisqacha rangli izohi tushuntirilib beriladi xaritalardagi kabi
# loc=0 bu qayerda joylashishini anglatadi. agar 1 qilsak tepa o'ng burchakda, 2 qilsak tepa chap burchakda joylashadi
# fontsize='x-large' bu uning hajmini belgilaydi. large kattaroq agar small qilsak kichikroq qilib chiqaradi
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.savefig("grafik1.png")  # bu bilan chap tomondagi faylga saqlab olib uni ochishimiz ham mumkin.
plt.show()
```

# **Marker Reference**


## Reference for filled-, unfilled- and custom marker types with Matplotlib.

 "Matplotlib yordamida to'ldirilgan, to'ldirilmagan va maxsus marker turlari uchun ma'lumotnoma."

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D


points = np.ones(3)  # Draw 3 points for each line
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontdict={'family': 'monospace'})
marker_style = dict(linestyle=':', color='0.8', markersize=10,
                    mfc="C0", mec="C0")


def format_axes(ax):
    ax.margins(0.2)
    ax.set_axis_off()
    ax.invert_yaxis()


def split_list(a_list):
    i_half = len(a_list) // 2
    return (a_list[:i_half], a_list[i_half:])
```

# **Filled and unfilled-marker types**
================================

## Plot all un-filled markers

To'ldirilgan va to'ldirilmagan marker turlari

Barcha to'ldirilmagan markerlarni chizish

```python
fig, axes = plt.subplots(ncols=2)
fig.suptitle('un-filled markers', fontsize=14)

# Filter out filled markers and marker settings that do nothing.
unfilled_markers = [m for m, func in Line2D.markers.items()
                    if func != 'nothing' and m not in Line2D.filled_markers]

for ax, markers in zip(axes, split_list(unfilled_markers)):
    for y, marker in enumerate(markers):
        ax.text(-0.5, y, repr(marker), **text_style)
        ax.plot(y * points, marker=marker, **marker_style)
    format_axes(ax)

plt.show()
```

# **Plot all filled markers.**

Barcha to'ldirilgan markerlarni chizish.

```python
fig, axes = plt.subplots(ncols=2)
for ax, markers in zip(axes, split_list(Line2D.filled_markers)):
    for y, marker in enumerate(markers):
        ax.text(-0.5, y, repr(marker), **text_style)
        ax.plot(y * points, marker=marker, **marker_style)
    format_axes(ax)
fig.suptitle('filled markers', fontsize=14)

plt.show()
```

# **Custom Markers with MathText**

"MathText yordamida maxsus markerlar"

```python
fig, ax = plt.subplots()
fig.subplots_adjust(left=0.4)

marker_style.update(mec="None", markersize=15)
markers = ["$1$", "$2$", "$f$", "$\\alpha$", "$\\beta$"]


for y, marker in enumerate(markers):
    # Escape dollars so that the text is written "as is", not as mathtext.
    ax.text(-0.5, y, repr(marker).replace("$", r"\\$"), **text_style)
    ax.plot(y * points, marker=marker, **marker_style)
format_axes(ax)
fig.suptitle('mathtext markers', fontsize=14)

plt.show()
```
