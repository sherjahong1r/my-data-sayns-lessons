# **.map() - qiymatlarni moslash (almashtirish)**

````python
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv', index_col=0)
df
````

````python
df['num-of-cylinders'].unique()
# num-of-cylinders ustunidagi noyob qiymatlarni chiqardik
````

````python
dict = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
# Bu qiymatlarning sondagi qiymatlari ham yozildi dict ga.
````

````python
df['num-of-cylinders'].map(dict)
# map yani moslashtirish bilan yuqoridagi sonli qiymatlar qo'yildi num-of-cylinders ustuniga.
````

````python
df['cylinders'] = df['num-of-cylinders'].map(dict)
df
# Yangi cylinders ustuniga saqlandi va bu ustun qo'shildi
````

````python
df['price']
# Bu yuqoridagi jadvalning price ustuni dollarda.
````

````python
kurs = 12000   # 1$ qancha ekanligi aniqlanib saqlandi
df['price_uz'] = df['price'].map(lambda x: x*kurs)
df
# lambda bir martalik nomsiz funksiyadan foydalanib, x ni qabul qilib yani price ustuni qiymatlarini
# kurs o'zgaruvchisiga yani 12000 ga ko'paytirib, so'ngra yangi price_uz ustuniga saqlanib chiqarayabdi.
````

````python
df[['num-of-cylinders','cylinders']]
````

`.map()` asosan Series (Pandas DataFrame'dagi bitta ustun) elementlarini birma-bir o'zgartirish uchun ishlatiladi. U ikkita asosiy holatda qo'llanilishi mumkin:

1.  **Lug'at (dictionary) yordamida:** Har bir qiymatni lug'atdagi kalitga qarab mos keladigan qiymatga almashtirish.
2.  **Funksiya yordamida:** Har bir qiymatga biror funksiyani (masalan, `lambda` funksiyasi) qo'llash.

### Notebookdagi misollar:

**1. Lug'at yordamida `num-of-cylinders` ustunini o'zgartirish:**

Siz `num-of-cylinders` ustunidagi 'four', 'six' kabi matnli qiymatlarni ularning sonli ekvivalentlariga o'zgartirdingiz. Buning uchun `dict` nomli lug'at yaratildi va keyin `map()` metodi bu lug'atni `num-of-cylinders` ustuniga qo'lladi.

```python
dict = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
df['cylinders'] = df['num-of-cylinders'].map(dict)
```

Bu kod `num-of-cylinders` ustunidagi har bir matnli qiymatni `dict` lug'atida mos keluvchi sonli qiymatga almashtirib, yangi `cylinders` ustunini yaratdi.

**2. `lambda` funksiyasi yordamida `price` ustunini o'zgartirish:**

Siz `price` ustunidagi dollar qiymatlarini o'zbek so'miga o'tkazish uchun `lambda` funksiyasidan foydalandingiz. `lambda` funksiyasi har bir `price` qiymatini `kurs` (12000) ga ko'paytirdi va natijalarni `price_uz` yangi ustuniga joylashtirdi.

```python
kurs = 12000
df['price_uz'] = df['price'].map(lambda x: x * kurs)
```

Bu kod `price` ustunidagi har bir qiymatni olib, uni `kurs` o'zgaruvchisiga ko'paytirdi va natijani `price_uz` ustuniga saqladi.
