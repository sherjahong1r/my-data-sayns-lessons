# **SUBPLOT**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df.shape
```

```python
df=df[df.year>=2010]             # 2010-yildan keyin chiqqan mashinlarni saqlab qoldik
df=df[df.transmission!='Other']  # transmission ustunidan Other degan qatorlarni tashlab yubordik
df=df[df.fuelType!='Other']      # fuelType ustunidan Other degan qatorlarni tashlab yubordik
df.head()
```

```python
df.shape
```

```python
x = np.arange(0,  4*np.pi, 0.1) # x nuqtasi 0 dan 4*pi gacha 0.1 oraliqda
sin = np.sin(x) # yuqoridagi x qiymatlarini sin qiymatida aniqladik sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6)) # 12 ga 6 hajmi
# sin(x) grafigi va ko'rinishini belgilovchi qiymatlar
plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         color='#8E44AD',
         alpha=0.7,        # bu shafoflik darajasi 0 dan 1 gacha agar 0 qilsak ko'rinmas edi.
         marker='h')       # nuqtalarini oltiburchak shaklida belgilash uchun

# cos(x) grafigi va ko'rinishini belgilovchi qiymatlar
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')      # pastga qaragan uchburchak bo'ladi

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x - o'qi nomi
plt.xlabel("pi o'qi")
# y - o'qi nomi
plt.ylabel("cos and sin o'qlar")
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x) funksiya','cos(x) funksiya'], loc=0, fontsize='x-large')
# legaend bilan sin va cos qisqacha rangli izohi tushuntirilib beriladi xaritalardagi kabi
# loc=0 bu qayerda joylashishini anglatadi. agar 1 qilsak tepa o'ng burchakda, 2 qilsak tepa chap burchakda joylashadi
# fontsize='x-large' bu uning hajmini belgilaydi. large kattaroq agar small qilsak kichikroq qilib chiqaradi
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.savefig("grafik1.png")  # bu bilan chap tomondagi faylga saqlab olib uni ochishimiz ham mumkin.
plt.show()
```

## Yuqoridagi 2 ta grafik bitta rasmda joylashgan edi
## subplot dan foydalanib ikkalasini ajratib qo'yamiz

```python
x = np.arange(0,  4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
# subplot(2,1,1) Bunda 2 degani ikkta qator qil, 1 degani bitta ustun qil, va yana 1 degani birinchi qatorga joylashtir degani.
plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         color='#8E44AD',
         alpha=0.7,
         marker='h')

plt.subplot(2,1,2)
# subplot(2,1,2) Bunda 2 degani ikkta qator qil, 1 degani bitta ustun qil, va yana 2 degani ikkinchi qatorga joylashtir degani.
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sin(x) va cos(x) funksiya grafigi')
plt.xlabel("pi o'qi")
plt.ylabel("cos and sin o'qlar")
plt.legend(['sin(x) funksiya','cos(x) funksiya'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.savefig("grafik1.png")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
# plt.subplots(2, 1, ...): Bu buyruq 2 qator va 1 ustundan iborat grafiklar maydonini yaratadi.
# fig butun rasm obyektini, ax esa har bir alohida grafik (subplot) uchun o'qlarni (axes) ifodalovchi massivni saqlaydi.
# figsize=(10,10): Butun rasmning hajmini 10x10 dyuym qilib belgilaydi.
sns.histplot(ax=ax[0], data=df, x='price')
# ax=ax[0]: Gistogrammani yaratilgan subplotlar ichidagi birinchi (yuqori) grafikga joylashtiradi.
sns.scatterplot(ax=ax[1], data=df, x='price', y='year', s=100)
# ax=ax[1]: Tarqalish diagrammasini yaratilgan subplotlar ichidagi ikkinchi (pastki) grafikga joylashtiradi.
# s=100: Har bir nuqtaning hajmini belgilaydi.
plt.show()
```

```python
fig1, ax1 = plt.subplots(2, 2, figsize=(15,10))
# fige1 ga saqlayabmiz bunda 2 ustun va 2 qator degani yani subplots(2,2)
fig1.suptitle("2 matrissali grafik")
# umumiy nom berildi
sns.histplot(ax=ax1[0,0], data=df, x='price')
# Bu grafik 0-ustun, 0-qatorga joylashayabdi
sns.histplot(ax=ax1[0,1], data=df, x='mileage')
# Bu grafik 0-ustun, 1-qatorga yani 0-qatordan pastga joylashayabdi
sns.scatterplot(ax=ax1[1, 0], data=df, x=df['year'], y=df['price'], hue="fuelType", s=100)
# Bu grafik 1-ustun yani 0-ustunni yoniga va 0-qatorga joylashadi
sns.scatterplot(ax=ax1[1, 1], data=df, x=df['year'], y=df['mileage'], hue="fuelType", s=100)
# Bu grafik 1-ustun va 1-qatorga joylashadi

# har bir grafiklarga nom berildi
ax1[0,0].set_title("price")
ax1[0,1].set_title("mileage")
ax1[1,0].set_title("year vs price")
ax1[1,1].set_title("year vs mileage")
plt.show()
```
