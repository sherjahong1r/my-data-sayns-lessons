## 8-VAZIFA. Uylarning narxi va xonalari soni uchun (2,3,5 xonali) KDE chizing.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head(1)
```

```python
df = df[df['rooms']>1]
df = df[df['rooms']<6]
df = df[df['rooms'] != 4]
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='rooms', multiple='stack', palette='Reds')
plt.show()

# 8 savolning 1 usuli lekin ko'p qatorli
```

```python
df = df[df['rooms'].isin([2, 3, 5])]
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='rooms', fill=True)
plt.show()
# 2 - onsonroq usuli
```

```python
#@title Kutilgan natija
sns.displot(data=df[df['rooms'].isin([2,3,5])], x='price', hue='rooms', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
# 3-usuli.
```

## 9-VAZIFA. Uylarning maydoni uchun kumulyativ taqsimot grafigini chizing (CDF)

```python
# JAVOB UCHUN JOY
sns.displot(data=df, x="size", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90) # x o'qidagi yillarni 90 gradusda joylashtiradi
plt.grid(which='both') # both ning o'rniga (major) kattaroq setka va (minor) turlari ham bor
plt.minorticks_on()
plt.show()

# plt.minorticks_on() grafigingizga qo'shimcha kichik o'q belgilarini qo'shib,
# ma'lumotlarni o'qishni osonlashtiradi va vizual aniqlikni oshiradi.
```

```python
#@title Kutilgan natija
sns.displot(data=df, x="size", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid() # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="size", hue='rooms', kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90) # x o'qidagi yillarni 90 gradusda joylashtiradi
plt.grid(which='major') # both ning o'rniga (major) kattaroq setka va (minor) turlari ham bor
plt.minorticks_on()
plt.show()
```
