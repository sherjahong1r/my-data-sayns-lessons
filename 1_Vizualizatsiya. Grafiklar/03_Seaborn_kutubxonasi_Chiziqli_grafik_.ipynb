# **Seaborn kutubxonasi. Chiziqli grafik.**

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns   # agar seaborn o'rnatilmagan bo'lsa pip install seaborn qilamiz
# boshqa kod muhitlari uchun. Shuningdek bu kutubxona grafiklarni chiroyliroq qilib chiqaradi
```

```python
dff = pd.read_csv('https://raw.githubusercontent.com/sherjahong1r/my-data-sayns-lessons/refs/heads/main/1_Vizualizatsiya.%20Grafiklar/tashkent_temp_csv', index_col=0)
dff.head()
# Bu bundan oldingi darsda ishlagan jadvalimiz.
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df)  # lineplot ga to'g'ridan to'g'ri data sifatida df ni uzatayabmiz
plt.show()
# Bu o'zi avtomatiskiy legend yani tushuntirish belgilarini qo'shib ketadi
# shuningdek xlabel yani x o'qi nomini ham qo'shib ketadi
```

```python
plt.plot(df)
plt.show()
# Bunda esa o'zi legendni yozmasak qo'shilmaydi
```

```python
plt.figure(figsize=(18,8))            # grafik hajmi
plt.title("Muzeylarga tashriflar")    # grafik nomi
plt.xlabel("Sana")                    # x o'qi nomi
plt.ylabel("Odamlar soni")            # y o'qi nomi
plt.xticks(rotation=90)
# yuqorida sanalar qora bo'lib ko'rinmay qolgandi bunda rotation=90 qilib kattalashtirdik va ko'ryabmiz
plt.grid()                            # bu setka chizadi
sns.lineplot(data=df)                 # legend berilayabdi sns kutubxonasi yordamaida
plt.show()                            # natijalar chiqarilayabdi
```

## ***Bunday grafiklarni qilishdan maqsad nima?***

Uzoq vaqt tavomidagi tahlillarni bitada qarab tushinish va qaysi oylarda, yillarda tashriflar soni oshgan yokida kamaygani bittada qarab tushinishimiz mumkin bo'ladi.

Yuqoridagi grafikda esa umumiy hisobda odamlarning tashrif soni kamayib borayotganini ko'rishimiz mumkin.

Shuningdek eng ko'p tashrif 2014 yil Firehouse Museum ga bo'lgan ekan.

Hamda 2015, 2017, 2018 yillarda Avila Adobe ga tashriflar tushgani va shu kabi boshqa jarayonlarni tushunish grafiklar orqali oson hisoblanadi.

```python
plt.figure(figsize=(18,8))
plt.title("Muzeylarga tashriflar")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(data=df[['Firehouse Museum', 'Chinese American Museum']])  # Bunda ikkta muzeyni o'zini ajratib oldik
plt.show()
```
