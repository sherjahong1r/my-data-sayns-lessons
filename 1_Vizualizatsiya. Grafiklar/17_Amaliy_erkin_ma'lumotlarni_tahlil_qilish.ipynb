```python
x, y = np.random.normal(size=(2, 4000)).cumsum(axis=1)
sns.lineplot(x=x, y=y, sort=False, lw=1)

# Bu kod parchasi numpy yordamida tasodifiy ma'lumotlarni yaratadi va seaborn yordamida ularni chiziqli grafik
# ko'rinishida vizualizatsiya qiladi. Keling, har bir qatorni ko'rib chiqaylik:

# x, y = np.random.normal(size=(2, 4000)).cumsum(axis=1): Bu kod tasodifiy yurish
# (random walk) usulida x va y koordinatalarini yaratadi, ya'ni tasodifiy sonlar asosida ketma-ketlik hosil qiladi.
# sns.lineplot(x=x, y=y, sort=False, lw=1): Bu kod Seaborn kutubxonasi yordamida
# x va y koordinatalari bo'yicha chiziqli grafik chizadi. sort=False chiziqlar yaratilgan
# tartibda chizilishini ta'minlaydi, lw=1 esa chiziq qalinligini belgilaydi.
```

```python
df = pd.read_excel("/kaggle/input/tashkent-real-estate-2019/uybor.xlsx")
df.head(1)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x="rooms", y="price", hue='level', s=100)
plt.show()
```

```python
sns.relplot(
    data=df, x="level", y="price",
    col="rooms", hue="rooms", style='level',
    kind="scatter"
)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.displot(data=df, x='price', height=10, aspect=2, hue='rooms', multiple='stack')
plt.show()
```

```python
df =df[df['rooms']<4]
sns.displot(data=df, x='price', height=5, aspect=2, col='rooms')
plt.show()
```

```python
sns.displot(data=df, x='price', height=5, aspect=2, kind='kde')
plt.show()
```

```python
sns.displot(data=df, x='price', height=5, aspect=2, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='rooms', shade=True)
plt.show()
```

```python
sns.displot(data=df, x="price", y="rooms", kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="price", hue="rooms", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
fig1, ax1 = plt.subplots(2, 2, figsize=(15,10))
# fige1 ga saqlayabmiz bunda 2 ustun va 2 qator degani yani subplots(2,2)
fig1.suptitle("2 matrissali grafik")
# umumiy nom berildi
sns.kdeplot(ax=ax1[0,0], data=df, x='price')

# Bu grafik 0-ustun, 0-qatorga joylashayabdi
sns.ecdfplot(ax=ax1[0,1],data=df, x="price", hue="rooms")

# Bu grafik 0-ustun, 1-qatorga yani 0-qatordan pastga joylashayabdi
sns.kdeplot(ax=ax1[1,0], data=df, x="price", y="rooms")

# Bu grafik 1-ustun yani 0-ustunni yoniga va 0-qatorga joylashadi
sns.heatmap(ax=ax1[1,1],data=df.corr(numeric_only=True), cmap='YlGnBu', linewidths=1, square=True)

# Bu grafik 1-ustun va 1-qatorga joylashadi

# har bir grafiklarga nom berildi
ax1[0,0].set_title("price KDE")
ax1[0,1].set_title("price ECDF by rooms")
ax1[1,0].set_title("price vs rooms KDE")
ax1[1,1].set_title("Correlation Heatmap")
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```
