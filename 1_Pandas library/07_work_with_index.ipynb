# **Index obyekti**

```python
import pandas as pd
```

```python
data = {
    'Yil':[2025, 2024, 2023, 2022, 2021, 2020, 2019, 2018],
    'Aholi soni':[36.5, 35.8, 35.2, 34.9, 34.1, 33.8, 33.0, 32.3],
    'Temp':[1.54, 1.48, 1.56, 1.62, 1.60, 1.59, 1.61, 1.57]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df = pd.DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YAIM']) # YAIM qiymati mavjud bo'lmagani uchun non deb chiqadi
```

```python
df
```

```python
df = pd.DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YAIM'], index=[25, 24, 23, 22, 21, 20, 19, 18]) # Hamda matnli index kiritsak ham bo'ladi bir, ikki deb.
df
```

```python
val = [1700, 1637, 1233, 1887, 1233, 1213, 1490, 1300]
df.YAIM = val
```

```python
df
```
