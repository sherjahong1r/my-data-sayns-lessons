```python
models = [' SLK',  'B Class']
filtr = df['model'].isin(models) # model dagi mashinalarni barchasini tekshir isin metodi bilan models = ro'yxatidagi qiymatlarini
filtr # Bor joylarini True, aks holda False qaytaradi
```

```python
df2 = df[filtr] # Bu bilan faqat (' SLK',  'B Class') shu mashinalar bor bo'lgan qatorlarni chiqaradi
df2 #  95 rows × 9 columns va bunda 95 qatordagi mashinalar (' SLK',  'B Class') rusumli ekan
```

```python
filtr = df['fuelType'].isin(['Diesel','Petrol'])
```

```python
df[filtr] # Bu orqali Diesel va Petrolda yuradigan mashinalr ro'yxatini shakllantirdik. 12939 ta qatordagi mashinalar shular orqali harakatlanadi
```

```python
df[df.year>2018] # 5_272 ta qatordagi mashinalar 2018 yildan keyin ishlab chiqarilgan ro'yxati
```

```python
df[df.year>=2018] # shu yildan katta yoki teng bo'lgan mashinalar ro'yxati.
```

```python
df[(df.year>=2020) & (df.price<25000)] # Bunda 2 ta shart bajarilyabdi yili 2020 dan katta yoki teng bo'lgan va
# narxi 25_000 $ dan past bo'lgan mashinalarni chiqaradi. & - va degani. Hamda bunday mashinalr 95 qatorda mavjud ekan
```

```python
df3 = df[(df.year>=2020) & (df.price<25000) & (df.mileage<80) & (df.fuelType == 'Diesel')] # Hohlagancha & bu bilan shartlarni qo'shish mumkin
df3 # Va df3 ga saqlab oldik
```
