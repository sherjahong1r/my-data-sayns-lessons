#### **`.insert()` Metodi**

`.insert()` metodi Series obyektiga belgilangan joyga (indexga) yangi element kiritish imkonini beradi. Bu metod yangi Series obyektini qaytaradi va asl Seriesni o'zgartirmaydi.

```python
import pandas as pd

data_df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print("Original DataFrame:")
display(data_df)

# 'C' ustunini 'B' ustunidan oldin (index 1) kiritish
data_df.insert(1, 'C', [7, 8, 9])
print("\nDataFrame after inserting column 'C' at index 1:")
display(data_df)

# Seriesga element 'kiritish' misoli (pd.concat orqali)
s = pd.Series([10, 20, 40])
print("\nOriginal Series:")
display(s)

# Index 2 ga '30' qiymatini kiritish uchun Seriesni qayta qurish yoki birlashtirish
s_inserted_series = pd.concat([s.iloc[:2], pd.Series([30]), s.iloc[2:]]).reset_index(drop=True)
print("\nSeries after 'inserting' 30 at index 2 (using pd.concat):")
display(s_inserted_series)

print("\nOriginal Series (unchanged after concat for Series):")
display(s)
```

#### **`is_monotonic` Metodi**

Bu metod Series yoki Index obyektining elementlari monoton o'suvchi (yoki bir xil) tartibda joylashganligini tekshiradi. Agar shunday bo'lsa `True`, aks holda `False` qaytaradi.

```python
import pandas as pd

monotonic_index = pd.Index([1, 2, 3, 5])
print(f"'monotonic_index' is monotonic increasing: {monotonic_index.is_monotonic_increasing}")
print(f"'monotonic_index' is monotonic decreasing: {monotonic_index.is_monotonic_decreasing}")

non_monotonic_index = pd.Index([1, 3, 2, 4])
print(f"'non_monotonic_index' is monotonic increasing: {non_monotonic_index.is_monotonic_increasing}")
print(f"'non_monotonic_index' is monotonic decreasing: {non_monotonic_index.is_monotonic_decreasing}")

decreasing_monotonic_index = pd.Index([5, 4, 3, 1])
print(f"'decreasing_monotonic_index' is monotonic increasing: {decreasing_monotonic_index.is_monotonic_increasing}")
print(f"'decreasing_monotonic_index' is monotonic decreasing: {decreasing_monotonic_index.is_monotonic_decreasing}")
```

#### **`is_unique` Metodi**

Bu metod Series yoki Index obyektidagi barcha elementlar noyob (takrorlanmaydigan) ekanligini tekshiradi. Agar shunday bo'lsa `True`, aks holda `False` qaytaradi.

```python
import pandas as pd
newindex = pd.Index(['olim','nafisa','hakim']) # newindex'ni shu yerda aniqlaymiz

print(f"'newindex' has unique values: {newindex.is_unique}")

duplicate_index = pd.Index(['a', 'b', 'a'])
print(f"'duplicate_index' has unique values: {duplicate_index.is_unique}")
```
