# **Dataset statistikasi: Korrelyasiya**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df # Onlayn bozordagi mersadesBens mashinalarining narxi
# Malumot uchun
# Model - nomi, year - yili, price - narxi, transmission - karobkasi, mileage - qancha yurgani, fuelType - nimada yurishi (Pertrol, Hybrid - elktr degani),
# tax - mashinaga beriladigan soliq, mpg - 1 galonga qancha km yurishi, engineSize - mashina matori hajmi 1.8 , 2.1 kabi.
```

```python
data = {
    'age':[10,20,30,40,50,60],
    'year':[2010,2000,1990,1980,1970,1960]
}
obj = pd.DataFrame(data)
obj
```

```python
# Korrelyasiya bu bog'liqlikni tekshiradi misol uchun age va year o'rtasidagi.
obj['age'].corr(obj['year']) # Bu korrelyasiya yani bog'liqlikni tekshiradi
# -0.999 yani bu deyarli 1 degani, shunda year va age o'rtasidagi bog'liqlik deyarli 100% ga teng degani.
```

```python
df['year'].corr(df['mileage']) # Bunda yil va bosib o'tgan klametri bir biriga bog'liq bo'lgani uchun year va mileage ni oldik
# -0.7 yani bu qanchalik 1 soniga yaqin bo'lsa shuncha bir biriga bog'liq yani korrelyasiya degani
```

```python
df['price'].corr(df['engineSize']) # mator va narx o'rtasidagi korrelyasiya. bu ham yaqin
```

```python
df.select_dtypes(include=np.number).corr() # df.corr() - desak ham bo'lardi ammo correlyasiya faqat sonli qiymatlarni oladi
# shuning uchun bu kod bilan sonli ustunlarni tanlab chiqaradi. Hamda hammasining bog'liqligini tekshiramiz
# Tahlil yearga nisbatan
# year + year 1 (korrelyasiya), year + price = 0.5, year + mileage = -0.7 bullarda bog'liqlik yuqori
# year + tax = 0.01, year + mpg = -0.09, year + engineSize = -0.1 bularda bog'liqlik kam shuning uchun natijasi 1 ga yaqin sonlar emas
```
