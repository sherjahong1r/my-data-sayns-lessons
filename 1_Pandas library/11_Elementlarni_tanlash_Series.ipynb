# **Indekslash, qator (ustun) tanlash, filtrlash**

```python
import pandas as pd
import numpy as np
```

```python
data = {"Yil":[2025, 2024, 2023, 2022, 2021, 2020, 2019,2014],
        "Aholi soni":[36.5, 35.4, 35.1, 34.3, 33.1, 32.4, 31.5, 30.8],
        "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        "Zichlik":[80.23, 79.45, 77.34, 75.13, 73.34, 71.23, 70.23, 68.23]}
df = pd.DataFrame(data)
df
```

```python
data = df[["Aholi soni"]]
data
```

```python
type(data) # Ikkita [[]] qavs qilsak DataFrame bo'ladi
```

```python
data = df["Aholi soni"]
data
```

```python
obj = pd.Series(np.arange(8.), index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
obj
```

```python
obj['b']
```

```python
obj[0:4] # 0 dan 4 qatorgacha chiqaradi
```

```python
obj['c':'f'] # c dan f gacha chiqaradi
```

```python
obj[['c','f']] # c va f ni chiqaradi
```

```python
obj
```

```python
obj[[1,5]]
```

```python
obj['a':'c']=15 # a dan c gacha 15 ga teng qildik
```

```python
obj
```

## **Filtrlash**

```python
obj[obj<15] # 15 dan kichiklarini chiqaradi
```

```python
obj[obj!=15] # 15 ga teng bo'lmaganlarni chiqaradi
```

```python
obj[obj==15] # 15 ga tenglarni chiqaradi
```

```python
obj[obj>5] # 5 dan kattalarni chiqaradi
```
