```python
import pandas as pd
import numpy as np
```

## Pandas Obyekti elementlariga Funksiyalarni qo'llash

```python
df = pd.DataFrame(np.random.randn(3, 4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
np.abs(-3) # Bu metod manfiy qiymatni musbat qiladi
```

```python
np.abs(df) # df dagi minus sonlar musbat bo'ladi
```

```python
np.round(df) # Bu metod sonlarni butunlab yani yaxlitlab beradi
```

```python
df
```

```python
df.mean() # ustunlar bo'yicha o'rtacha qiymatni chiqaradi
```

```python
df.mean(axis=1) # qatorlar bo'yicha o'rtacha qiymat
```

```python
df.apply(np.abs)
```

```python
df.apply(sum) # Bu ham o'rtacha qiymatni chiqaradi Ustunlar bo'yicha
```

```python
df.apply(sum, axis=1) # Qatorlar bo'yicha
```

```python
f = lambda x: x.max() - x.min() # Ustunlardagi maxsimum son va  minimum son o'rtasidagi farqni chiqaradi
```

```python
df.apply(f)
```

```python
def foo(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo)
```

```python
df = pd.DataFrame(np.arange(1,13).reshape(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
df.applymap(lambda x: x*x) # Bu x qiymat qabul qiladi va uni o'ziga o'zini ko'paytirib yani shu sonni kvadratini chiqaradi
```
