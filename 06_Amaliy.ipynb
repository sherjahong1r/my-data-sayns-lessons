```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#JAVOBINGIZNI SHU YERGA YOZING
df = pd.read_csv("https://raw.githubusercontent.com/sherjahong1r/my-data-sayns-lessons/refs/heads/main/1_Vizualizatsiya.%20Grafiklar/tashkent_temp_csv", index_col=0)
df.head()
```

### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)


T	Average annual temperature (C)

TM	Annual average maximum temperature (C)

Tm	Average annual minimum temperature (C)

PP	Rain or snow precipitation total annual

V	Annual average wind speed (km/h)

RA	Number of days with rain

SN	Number of days with snow

TS	Number of days with storm

FG	Number of foggy days

TN	Number of days with tornado

GR	Number of days with hail

T  Ortacha_harorat (C)

TM  Max_harorat (C)

Tm  Min_harorat (C)

PP  Yogin_miqdori (yiliga jami)

V  Shamol_tezligi (km/soat)

RA  Yomgir_kunlari

SN  Qor_kunlari

TS  Borron_kunlari

FG  Tuman_kunlari

TN  Tornado_kunlari

GR  Dol_kunlari

```python
columns = ["Ortacha_temp","Max_temp","Min_temp","Shamol_tezligi","Yomgir_kunlari","Qor_kunlari","Bo'ron_kunlari","Tuman_kunlari","Tornado_kunlari","Do'l_kunlari"]
df.columns = columns
df.head()
```

```python
plt.figure(figsize=(14,5))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df, linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

### 4-VAZIFA. Yuqoridagi grafikda turli ko'rsatkichlar aralashib ketdi. Keling ustun turiga qarab alohida grafiklar chizamiz.

#### 4.1. Haroratni ko'rsatuvchi ustunlarni alohida chiziqli grafik qilib chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(18,8))
plt.title("Ob havo ko'rsatkichlari Toshkent shaxri uchun")
plt.xlabel("Yili")
plt.ylabel("Harorat")
#plt.xticks(rotation=90) Bu vertikal yani uzunasiga joylashtiradi
plt.grid()
sns.lineplot(data=df[['Ortacha_temp', 'Max_temp','Min_temp']])  # Bunda ikkta muzeyni o'zini ajratib oldik
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df[['Ortacha_temp','Max_temp','Min_temp']], linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

#### 4.2. Kunlarni ko'rsatuvchi ustunlarni alohida grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(18,8))
plt.title("Ob havo ko'rsatkichlari Toshkent shaxri uchun")
plt.xlabel("Yili")
plt.ylabel("Kunlar kesimida")
#plt.xticks(rotation=90) Bu vertikal yani uzunasiga joylashtiradi
plt.grid()
sns.lineplot(data=df[['Yomgir_kunlari', 'Qor_kunlari',"Bo'ron_kunlari","Tuman_kunlari","Tornado_kunlari","Do'l_kunlari"]])  # Bunda ikkta muzeyni o'zini ajratib oldik
plt.show()
```

```python
#@title Kutilgan natija:
labels=[
        "Yomg'irli kunlar soni",
        "Qorli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df[['Yomgir_kunlari','Qor_kunlari','Bo\'ron_kunlari','Tuman_kunlari','Tornado_kunlari']], linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Kunlar")
plt.legend(labels, loc=1)
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.
