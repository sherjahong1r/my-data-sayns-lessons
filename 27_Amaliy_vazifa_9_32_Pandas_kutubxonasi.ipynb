## `DataFrame: Asoslar`

DataFrames始da ma始lumotlarni tanlash (selecting), saralash (sorting), qo驶shish (adding) va yig驶ish (aggregating) uchun bir nechta asosiy mashqlar.

Eslatma: NumPy kutubxonasini import qilishni unutmang.

Quyidagi Python lug驶ati (dictionary) `data` va Python ro驶yxati (list) `labels` ni ko驶rib chiqing.


``` python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],

```



````python
import numpy as np
import pandas as pd

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# DataFrame yaratish
df = pd.DataFrame(data, index=labels)

# Natijani chiqarish
print(f"Bu umumiy jadvalning ko'rinishi:\n\n{df}")

print("\n1) Bu tanlash (selecting):\n")
print(df[['animal', 'age']])
print("\nIndex bo'yicha tanlash\n")
print(df.iloc[1]) # b qatordagi barcha ustunlarni chiqaradi
print("\nShart bo'yicha tanlash\n")
print(df[df['age']>3]) # age dagi 3 yoshdan kattalarni chiqaradi

print("\n2) saralash (sorting):\n")
print(df.sort_values(by="age")) # yoshni o'sish bo'yicha saralaydi
print("\nYuqoridagi kichikdan kattaga, pastdagi kattadan kichikga sorting qiladi yoshga nisbatan\n")
print(df.sort_values(by="age", ascending=False)) # yoshni kamayish bo'yicha saralaydi

print("\n3) qo'shish (adding):\nNew gender ustun added\n")
df['gender'] = ['male', 'male', 'male', 'female', 'female', 'female', 'male', 'male', 'male', 'male']
print(df)

print("\n4) yig驶ish (aggregating):\n")
print("\nYoshning o'rtacha qiymati hisoblandi")
print(df['age'].mean())
print("\nmax and min qiymatlar\n")
print(df['age'].max)
print("\n")
print(df['age'].min)
print("\nHar bir turdagi animal soni hisoblandi\n")
print(df['animal'].value_counts())
````

**31**. `RecentDelays` ustunida qiymatlar DataFrame始ga ro驶yxat sifatida kiritilgan. Vazifangiz birinchi, har bir ikkinchi qiymatni o驶z ustunida va qolgan qiymatlar ham ustunga mos holda chop etish. Agar N-chi qiymat bo驶lmasa, qiymat `NaN` deb nomlanishi kerak.

Ro驶yxatlar qatorini `delays` nomli DataFrame ichiga kengaytiring, `delay_1`, `delay_2` va qolgan ustunlarni ham nomlarini o驶zgartiring va `df` dagi keraksiz `RecentDelays` ustunini `delays` bilan almashtiring.

Natija quyidagi ko驶rinishda bo驶lishi kerak:
```
   FlightNumber          Airline      From         To  delay_1  delay_2  delay_3
0         10045              KLM    London      Paris     23.0     47.0      NaN
1         10055       Air France    Madrid      Milan      NaN      NaN      NaN
2         10065  British Airways    London  Stockholm     24.0     43.0     87.0
3         10075       Air France  Budapest      Paris     13.0      NaN      NaN
4         10085        Swiss Air  Brussels     London     67.0     32.0      NaN
```

````python
import numpy as np
import pandas as pd

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
                               'Budapest_PaRis', 'Brussels_londOn'],
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
                               '12. Air France', '"Swiss Air"']})
display(f"{df}\n")
df.loc[1, 'FlightNumber'] = 10055
df.loc[3, 'FlightNumber'] = 10075
df['FlightNumber'] = df['FlightNumber'].astype(int)

df['Airline'] = df['Airline'].str.replace(r'[^a-zA-Z\s]', '', regex=True).str.strip()

temp = df['From_To'].str.split('_', expand=True)
temp.columns = ['From', 'To']

temp['From'] = temp['From'].str.capitalize()
temp['To'] = temp['To'].str.capitalize()

df = pd.concat([df.drop('From_To', axis=1), temp], axis=1)

###############
delays = df['RecentDelays'].apply(pd.Series)
delays.columns = [f'delay_{i+1}' for i in range(delays.shape[1])]
df = pd.concat([df.drop('RecentDelays', axis=1), delays], axis=1)
display(df)
````
