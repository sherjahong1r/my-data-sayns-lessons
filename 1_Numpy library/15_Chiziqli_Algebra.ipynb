### Qo'shimcha Matritsa Ko'paytirish Misollari

**1-misol:** 1x3 matritsani 3x1 matritsaga ko'paytirish.

```python
a = np.array([[1, 2, 3]]) # 1x3 matritsa
b = np.array([[4], [5], [6]]) # 3x1 matritsa

print('a matritsasi:')
print(a)
print('\nb matritsasi:')
print(b)

# Natija 1x1 matritsa bo'ladi
result1 = np.dot(a, b)
print('\na * b (np.dot) natijasi:')
print(result1)
```

**2-misol:** 2x2 matritsani 2x2 matritsaga ko'paytirish.

```python
c = np.array([[1, 0], [0, 1]]) # 2x2 birlik matritsa
d = np.array([[2, 3], [4, 5]]) # 2x2 matritsa

print('c matritsasi:')
print(c)
print('\nd matritsasi:')
print(d)

# Natija 2x2 matritsa bo'ladi
result2 = np.dot(c, d)
print('\nc * d (np.dot) natijasi:')
print(result2)

# '@' operatori bilan ham shu ishni qilish mumkin
result2_at = c @ d
print('\nc @ d natijasi:')
print(result2_at)
```

**3-misol:** 2x3 matritsani 3x2 matritsaga ko'paytirish.

```python
e = np.array([[1, 2, 3], [4, 5, 6]]) # 2x3 matritsa
f = np.array([[7, 8], [9, 10], [11, 12]]) # 3x2 matritsa

print('e matritsasi:')
print(e)
print('\nf matritsasi:')
print(f)

# Natija 2x2 matritsa bo'ladi
result3 = np.dot(e, f)
print('\ne * f (np.dot) natijasi:')
print(result3)
```
