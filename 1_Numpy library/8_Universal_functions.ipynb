# Universal funksiyalar

## Unary funksiyalar (bitta argument qabul qiluvchi funksiya)

## sqrt, square, exp, log, modf, sign, isnan funksiyalarini ko'rib chiqamiz

```python
arr = np.arange(10) # 0 ~ 10 gacha bo'lgan elementlardan iborat massivni qaytaradi
arr
```

## sqrt massivning xar bir elementini kvadrat ildizini qaytaradi

```python
np.sqrt(arr) # arr nomli massivning barcha elementini kvadratga oshiradi. Hamda unary bo'lgani uchun bitta o'zgaruvchi qiymatini oladi
```

## square massivning xar bir elementini kvadratga oshiradi

```python
np.square(arr)# arr nomli massivning barcha elementini kvadratga oshiradi
```

## exp massivning xar bir elementini kvadratga oshiradi

```python
np.exp(arr) # arr nomli massivning barcha elementlarini eksponentini qaytaradi
```

```python
arr
```

## log massivning xar bir logorifimini hisoblaydi

```python
np.log(arr) # arr nomli massivning barcha elementini logarifimini hisoblaydi(loga)
```

```python
np.log(arr[1:]) # Bu bilan 0 xato bergani uchun qaytadan 1 dan boshladik
```

## modf bu funksiya haqiqiy sonlarni butun va qoldiq qismlarini ajratib beruvchi funksiya

```python
qoldiq, butun = np.modf(arr1) # arr1 ning elementlarini qoldiq va butun qismlarini ikkita massivga ajratadi
```

```python
print(qoldiq)
print(butun)
```

```python
np.sign(arr1) # arr1 massivining elementlari ishoralarini "-1" va "1" ko'rinishida qaytaradi
```

## abs va fabs funksiyalari (absolyut qiymat)

```python
arr_abs_fabs = np.array([-1.2, 0.5, -3, 4, -0.0])
print(f"Original array for abs/fabs: {arr_abs_fabs}")
```

### abs: massiv elementlarining absolyut qiymatini hisoblaydi

```python
print(f"np.abs(arr_abs_fabs): {np.abs(arr_abs_fabs)}")
```

### fabs: massiv elementlarining absolyut qiymatini float ko'rinishida hisoblaydi

```python
print(f"np.fabs(arr_abs_fabs): {np.fabs(arr_abs_fabs)}")
```

## log10 funksiyasi (10 asosli logarifm)

```python
arr_log10 = np.array([1, 10, 100, 1000])
print(f"Original array for log10: {arr_log10}")
```

### log10: massiv elementlarining 10 asosli logarifmini hisoblaydi

```python
print(f"np.log10(arr_log10): {np.log10(arr_log10)}")
```

## ceil, floor, rint funksiyalari (yaxlitlash)

```python
arr_round_new = np.array([-1.7, -1.5, -1.2, 0.5, 1.2, 1.5, 1.7])
print(f"Original array for rounding: {arr_round_new}")
```

### ceil: massiv elementlarini yuqoriga yaxlitlaydi (ceiling)

```python
print(f"np.ceil(arr_round_new): {np.ceil(arr_round_new)}")
```

### floor: massiv elementlarini pastga yaxlitlaydi (floor)

```python
print(f"np.floor(arr_round_new): {np.floor(arr_round_new)}")
```

### rint: massiv elementlarini eng yaqin butun songa yaxlitlaydi

```python
print(f"np.rint(arr_round_new): {np.rint(arr_round_new)}")
```

## modf funksiyasi (butun va qoldiq qismlarini ajratish)

```python
arr_modf = np.array([-3.7, -1.5, 0.9, 2.3])
print(f"Original array for modf: {arr_modf}")
qoldiq_new, butun_new = np.modf(arr_modf)
print(f"Qoldiq (fractional part): {qoldiq_new}")
print(f"Butun (integer part): {butun_new}")
```

### sin: massiv elementlarining sinusini hisoblaydi

### sinh: massiv elementlarining giperbolik sinusini hisoblaydi

### tan: massiv elementlarining tangensini hisoblaydi

```python
print(f"np.tan(arr_trig_new): {np.tan(arr_trig_new)}")
```

### tanh: massiv elementlarining giperbolik tangensini hisoblaydi

```python
print(f"np.tanh(arr_trig_new): {np.tanh(arr_trig_new)}")
```

## Binary funksiyalar (ikkita argument qabul qiluvchi funksiyalar)

```python
# Misollar uchun ikkita massiv yaratib olamiz
arr_a = np.array([10, 20, 30, 40, 50])
arr_b = np.array([3, 5, 2, 8, 10])
print(f"arr_a: {arr_a}")
print(f"arr_b: {arr_b}")
```

### subtract: ikkita massivning mos elementlarini ayiradi

```python
print(f"np.subtract(arr_a, arr_b): {np.subtract(arr_a, arr_b)}")
```

### power: birinchi massiv elementlarini ikkinchi massiv elementlari darajasiga oshiradi

```python
print(f"np.power(arr_a, arr_b): {np.power(arr_a, arr_b)}")
```

### maximum: ikkita massivning mos elementlaridan kattasini qaytaradi

```python
print(f"np.maximum(arr_a, arr_b): {np.maximum(arr_a, arr_b)}")
```

### fmax: ikkita massivning mos elementlaridan kattasini qaytaradi, NaN qiymatlarni e'tiborsiz qoldiradi

```python
arr_c = np.array([1, np.nan, 3, 4])
arr_d = np.array([np.nan, 2, 1, 5])
print(f"arr_c: {arr_c}")
print(f"arr_d: {arr_d}")
print(f"np.fmax(arr_c, arr_d): {np.fmax(arr_c, arr_d)}")
```

### minimum: ikkita massivning mos elementlaridan kichigini qaytaradi

```python
print(f"np.minimum(arr_a, arr_b): {np.minimum(arr_a, arr_b)}")
```

### fmin: ikkita massivning mos elementlaridan kichigini qaytaradi, NaN qiymatlarni e'tiborsiz qoldiradi

```python
print(f"np.fmin(arr_c, arr_d): {np.fmin(arr_c, arr_d)}")
```

### mod: ikkita massivning mos elementlari bo'linmasining qoldig'ini qaytaradi

```python
print(f"np.mod(arr_a, arr_b): {np.mod(arr_a, arr_b)}")
```

### copysign: birinchi massiv elementlarining absolyut qiymatiga ikkinchi massiv elementlarining ishorasini qo'yadi

```python
arr_e = np.array([1.0, -2.0, 3.0, -4.0])
arr_f = np.array([-1.0, 2.0, -3.0, 4.0])
print(f"arr_e: {arr_e}")
print(f"arr_f: {arr_f}")
print(f"np.copysign(arr_e, arr_f): {np.copysign(arr_e, arr_f)}")
```

## Taqoslash funksiyalari (Comparison functions)

```python
arr_g = np.array([1, 2, 3, 4, 5])
arr_h = np.array([2, 2, 1, 4, 6])
print(f"arr_g: {arr_g}")
print(f"arr_h: {arr_h}")
```

### greater: birinchi massiv elementi ikkinchidan katta bo'lsa True qaytaradi

```python
print(f"np.greater(arr_g, arr_h): {np.greater(arr_g, arr_h)}")
```

### greater_equal: birinchi massiv elementi ikkinchidan katta yoki teng bo'lsa True qaytaradi

```python
print(f"np.greater_equal(arr_g, arr_h): {np.greater_equal(arr_g, arr_h)}")
```

### less: birinchi massiv elementi ikkinchidan kichik bo'lsa True qaytaradi

```python
print(f"np.less(arr_g, arr_h): {np.less(arr_g, arr_h)}")
```

### less_equal: birinchi massiv elementi ikkinchidan kichik yoki teng bo'lsa True qaytaradi

```python
print(f"np.less_equal(arr_g, arr_h): {np.less_equal(arr_g, arr_h)}")
```

### equal: ikkala massiv elementlari teng bo'lsa True qaytaradi

```python
print(f"np.equal(arr_g, arr_h): {np.equal(arr_g, arr_h)}")
```

### not_equal: ikkala massiv elementlari teng bo'lmasa True qaytaradi

```python
print(f"np.not_equal(arr_g, arr_h): {np.not_equal(arr_g, arr_h)}")
```

## Mantiqiy funksiyalar (Logical functions)

```python
arr_bool1 = np.array([True, False, True, False])
arr_bool2 = np.array([True, True, False, False])
print(f"arr_bool1: {arr_bool1}")
print(f"arr_bool2: {arr_bool2}")
```

### logical_and: ikkala element ham True bo'lsa True qaytaradi

```python
print(f"np.logical_and(arr_bool1, arr_bool2): {np.logical_and(arr_bool1, arr_bool2)}")
```

### logical_or: elementlardan hech bo'lmaganda bittasi True bo'lsa True qaytaradi

```python
print(f"np.logical_or(arr_bool1, arr_bool2): {np.logical_or(arr_bool1, arr_bool2)}")
```

### logical_xor: elementlardan faqat bittasi True bo'lsa True qaytaradi

```python
print(f"np.logical_xor(arr_bool1, arr_bool2): {np.logical_xor(arr_bool1, arr_bool2)}")
```
