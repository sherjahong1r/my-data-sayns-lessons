```python
import numpy as np
```

# **Massivga yo'naltirilgan dasturlash**

## Tartiblash (Sorting)

```python
# data
arr = np.random.rand(4) # 4 ta taxminiy elementlardan tashkil topgan massiv
arr
```

## sort funksiyasi

```python
# 1 usul
sorted_arr = np.sort(arr) # arr massivining tartiblangan versiyasi
sorted_arr # Bu buytuq bilan kichikdan katta raqamga o'sib boradi
```

```python
# 2 usul
arr.sort() # arr ning tartiblangan ko'rinishi
arr
```

```python
# Teskari tartiblash yani kattadan kichikga kamayib boradi
rev_arr = -np.sort(-arr) # teskari tartibdagi arr
rev_arr
```

```python
# 2 D array (massiv)
arr2d = np.random.randn(2, 4) # (2, 4) random massiv
arr2d
```

```python
# axis = 0 --> Qator
row_sort = np.sort(arr2d, axis=0) # arr2d ning xar bir qator bo'yicha tartiblangan ko'rinishi yani vertikal bo'yicha.
row_sort
```

```python
# axis = 1 --> Ustun yani gorizontal tartibda
col_srt = np.sort(arr2d, axis=1) # arr2d ning xar bir ustun bo'yicha tartiblangan ko'rinishi yani gorizantal
col_srt
```

```python
# Bu pastagi kattadan kichikga qarab boradi
```

```python
col_srt = -np.sort(-arr2d, axis=1) # arr2d ning xar bir ustun bo'yicha tartiblangan ko'rinishi yani gorizantal
col_srt
```
