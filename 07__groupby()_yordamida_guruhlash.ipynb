# **.groupby() yordamida guruhlash**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df
```

```python
df.company.value_counts()
# company ustunidagi takrorlanmas mashinalar va ularning soni aniqlandi.
```

```python
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].mean(numeric_only=True)
```

```python
df.groupby('company').mean(numeric_only=True)
# Bu ham ustun bo'yicha amallar bajarishning yana bir usuli.
# numeric_only=True bilanfaqat sonli ustunlarni tanlab olyabmiz.
```

```python
df.groupby('company').price.max()
# takrorlanmas mashinalar uchun maxsimum narxlar chiqarildi.
```

```python
df.groupby(['company','body-style']).price.mean()
# company va body-style ustunlarni chiqardik hamda price bo'yicha ham o'rtacha qiymatini chiqardik.
# alfa-romero companiyasining 2 ta yani convertible va hatchback rusumli mashinlari hisoblangan, qolganlari ham shu kabi.
```

```python
#df.groupby('company').agg([min, max, np.mean])


df.groupby('company').agg({'wheel-base': ['min', 'max', 'mean', lambda x: np.median(x)],
                               'length': ['min', 'max', 'mean', lambda x: np.median(x)],
                               'horsepower': ['min', 'max', 'mean'],
                               'average-mileage': ['min', 'max', 'mean'],
                               'price': ['min', 'max', 'mean', lambda x: np.median(x)]})

# agg - agregatik, yani bir nechta buyruqlarini bittada yozish mumkin
# bunda biz tanlagan guruhlar bo'yicha min, max va mean qiymatlarini hisoblab beradi.
# Shuningdek ayrim ustunlarga lambda x: np.median(x) bu buyruqdan ham foydalanganmiz yani bu lambda bilan medianini hisoblab beradi.
```

```python
df.groupby('company').price.agg([min, max, np.mean, lambda x: np.median(x)])
# Bunda 1 ta ustun yani price bo'yicha chiqarildi.
```

```python
df.groupby('company').agg([len, min, max])
```

```python
df.groupby('body-style').price.agg([len, min, max, np.mean])
```
