## Mantiqiy shart operatorlarni massiv operator sifatida qo'llash

## Shart operatori sifatida where dan foydalanish

## Tassavur qiling bizda ikkita ma'lumotlar (massivlar) mavjud bular: xarr va yarr. Agarda shart True ni qanoatlantirsa xarr elementini qabul qilgan aksincha esa yarr elementini qabul qilgan yangi massiv hosil qilinsin

```python
# Massivlar
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5]) # xarr
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5]) # yarr
```

```python
# Shart
cond = np.array([True, False, True, True, False])
```

```python
# where
result = np.where(cond, xarr, yarr) # cond==True : xarr elementi, cond==False : yarr elementini chiqaradi, va buni np.where buyrug'i orqali qilamiz.
result
```

## New example with where

```python
arr = np.random.randn(4, 4) # (4, 4) taxminiy massiv
arr
```

```python
# where
results = np.where(arr<0, -2, 2) # arr ning manfiy elementlarini -2 ga musbat elementlarini esa 2 ga almashtirildi
results
```

## New example with where

```python
ism = np.array(['shokir', 'tora', '2004 - ', 'dimka', 'tugilgan'])
familya = np.array(['qosimov', 'abduvaliyev', 'kunda', 'yilda', 'ha'])
```

```python
shart = np.array([True, False, True, False, True])
```

```python
natija = np.where(shart, ism, familya)
print(natija)
```
