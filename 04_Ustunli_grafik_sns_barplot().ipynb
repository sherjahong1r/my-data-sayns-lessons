# **Ustunli grafik. sns.barplot()**

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv', index_col=0)
df.columns = ['maydon', 'aholi']
df
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y=df['aholi'])
# x o'qiga df jadvalining index o'rnida turgan viloyatlar nomini qo'yayabmiz
# y ga ustun qilib esa aholi ustuni qiymatlarini qo'ydik

plt.title("O'zbekistonda hududlar kesimidagi aholi soni")
plt.xlabel("Hudud")
plt.ylabel("Aholi soni (million)")
plt.grid()  # gorizontal va vertikal setka tortish
plt.xticks(rotation=98)
plt.show()
```

```python
# Bu maydoni bo'yicha grafigi

plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y=df['maydon'])

plt.title("O'zbekistonda hududlar kesimidagi maydon taqsimoti")
plt.xlabel("Hudud")
plt.ylabel("maydon (kv.km)")
plt.grid()
plt.xticks(rotation=98)
plt.show()
```
