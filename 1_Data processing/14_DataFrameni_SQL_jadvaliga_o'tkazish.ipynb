# **Kerakli modullarni chaqirib olamiz**

```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("new.db") # bu new.db MO ini db deb saqladik
cursor = db.cursor()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.head()
```

```python
cols = ["country", "code", "year", "coal", "gas", "hydro", "renewable_other", "solar", "oil", "wind", "nuclear"]
df.columns = cols
df.head()
# Yuqoridagi jadvalni ustun nomlarini qaytadan qisqartirib saqladik.
```

```python
df.wind.head()
```

```python
df.country.tail()
```

```python
df.country.sample(5)
```

```python
df.to_sql("electric_45", db, if_exists='replace')
# df jadvali eng yuqoridagi db omboriga electric_45 deb saqladik saqlandi sql formatda.
# if_exists='replace bu agar shunday nomli jadval bo'lsa uni almashtirib qaytadan saqlaydi
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
# Bu bilan bunday jadvalimiz mavjudligini ko'ramiz
```

```python
cursor.execute("PRAGMA table_info(electric_45);")
cursor.fetchall()
# Bu jadval ustunlarini ko'ramiz
```

```python
query = "SELECT * FROM electric_45"
df = pd.read_sql(query, db)
df.head()
```

```python
query = "SELECT * FROM electric_45 WHERE country='Uzbekistan'"
pd.read_sql(query, db)
```
