```python
import sqlite3

# 1) Bazaga ulanish (timeout bilan)
with sqlite3.connect("example.db", timeout=20) as db:
    cursor = db.cursor()

    # 2) Jadvalni yaratish
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS worldPopulation (
        rank INTEGER,
        country TEXT,
        density REAL,
        population INTEGER,
        area INTEGER
    )
    """)


    # 3) Ma'lumot qoshish
    cursor.execute("""
    INSERT INTO worldPopulation (country, density, population, area)
    VALUES ('Kazakhstan', 6.9, 19000000, 2725000)
    """)

    # 4) Natijani korish
    cursor.execute("SELECT * FROM worldPopulation")
    rows = cursor.fetchall()
print("\n")
# 5) Chiqarish
for row in rows:
    print(row)



query = "INSERT INTO worldPopulation VALUES(1, 'Uzbekistan', 33000, 33000000, 445000)"
cursor.execute(query)
db.commit()
# Ustunlarga mos VALUES bilan qiymatlar kiritildi.
```
