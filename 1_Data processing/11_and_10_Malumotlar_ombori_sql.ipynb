```python
db = sql.connect("chinook.db")  # Malumotlar omboriga db deb nom berdik
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
# Jadvallar ro'yxati
```

```python
cursor.execute("SELECT COUNT (*) FROM employees")
cursor.fetchone()
# employesda nechta odam borligi
```

```python
df = pd.read_sql_query("SELECT * FROM customers LIMIT 10", db)
df
# df ga saqlab olib data frame ko'rinishida jadvalga chiqarayabmiz
# data frame o'zi index berib ketdi
# LIMIT 10 bilan boshidagi 10 qatorini chiqaradi
```

```python
df = pd.read_sql_query("SELECT * FROM customers LIMIT 5", db, index_col='CustomerId')
df
# Bunda index_col bilan CustomerId ustunini index o'riga qo'ydik
```

```python
df = pd.read_sql_query("SELECT CustomerId, Firstname, LastName, Phone FROM customers", db, index_col='CustomerId')
df.head()
# Yuqorida * bo'lgani uchun hamma ustunni olgan edi.
# Ammo biz * o'rniga faqat biz tanlagan ustunlarni ham chiqarsak bo'ladi.
```
