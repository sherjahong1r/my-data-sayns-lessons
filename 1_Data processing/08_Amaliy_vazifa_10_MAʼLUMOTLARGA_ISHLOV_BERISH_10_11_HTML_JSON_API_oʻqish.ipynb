#1-topshiriq
1. Pandas kutubxonasidan foydalanib, quyida berilgan link ichidan barcha jadvallarni o‘qing va roʻyxat shaklida saqlang.

Link 403 Forbidden (Manzil to'g'ri, lekin kirish taqiqlangan.) xatoligini bermasligi uchun, sahifani avval requests bilan yuklab, keyin pd.read_html() ga HTML matnini bering.

 * https://en.wikipedia.org/wiki/Fossil_fuel_exporters
2. Har bir jadvalning birinchi 1 qatorini ekranga chiqaring.


    

```python
import pandas as pd
import requests
from io import StringIO # Import StringIO

url = "https://en.wikipedia.org/wiki/Fossil_fuel_exporters"

# User-Agent sarlavhasini qo'shamiz
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)

# Sahifa kontentini StringIO orqali pd.read_html ga beramiz
tables = pd.read_html(StringIO(response.text)) # Use StringIO here

#Har bir jadvalning birinchi 1 qatorini ekranga chiqaring.
# for i, table in enumerate(tables):
#     print(f"Jadval {i+1} ning birinchi qatori:")
#     display(table.head(1))
#     print("\n")

a = len(tables)
print(f"Umumiy ({a}) ta jadval bor ekan\n")
b = tables[1].head(1)
print(f"{b}\n")
c = tables[2].head(1)
print(f"{c}")
```

Yuqoridagi HTML kontentni ko'rib chiqing. Agar u erda `<table>` teglari bo'lmasa, `pd.read_html` ularni topa olmaydi. Agar `<table>` teglari mavjud bo'lsa, unda muammo `pd.read_html` ning parslash usulida bo'lishi mumkin va `beautifulsoup4` kabi boshqa kutubxonadan foydalanish kerak bo'ladi.

#2-topshiriq
Yuqorida yaratilgan jadvallar roʻyxatidan eng uzunini `largest_data.csv` fayliga saqlang.

```python
largest_table = max(tables, key=len)
max_rows = len(largest_table)

print(f"Eng uzun jadvalda {max_rows} qator bor.")

largest_table.to_csv('largest_data.csv', index=False)
print("Eng uzun jadval 'largest_data.csv' fayliga saqlandi.")
```

#4-topshiriq
1. Quyida berilgan linkga `get` soʻrovini yuborish orqali `json` maʼlumotlarini `posts.json` fayliga saqlang.
  * https://jsonplaceholder.typicode.com/posts/
2. Ushbu maʼlumotlarni dataframe koʻrinishida saqlang va tasodifiy 5 ta qatorni ekranga chiqaring.

```python
import pandas as pd

url = "https://jsonplaceholder.typicode.com/posts/"

df_posts = pd.read_json(url)

df_posts.to_json('posts.json', indent=4, orient='records')

print("Ma'lumotlar 'posts.json' fayliga saqlandi.")
print("Ma'lumotlar dataframe ko'rinishida:")
display(df_posts.sample(5))
```
