# **JSON va API'dan o'qish**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv")
df
```

```python
hdfobj = pd.HDFStore("mydata.h5")
# mydata.h5 degan folder ochdik
```

```python
hdfobj.put("kompany",df)
# Hamda mydata.h5 ichiga hdfobj bilan yuqoridagi df ro'yxatni kompany deb saqladik
```

```python
hdfobj.keys()
# mydata.h5 ichidagi fayllar ro'yxati
```

```python
hdfobj['kompany'].head()
# mydata.h5 ichidagi kompany ro'yxatini boshidagi 5 qatorini chiqardik
```

```python
df.to_json("newkompany.json")
df
# Eng yuqoridagi df jadvalini json formatiga o'tkazdik
```
