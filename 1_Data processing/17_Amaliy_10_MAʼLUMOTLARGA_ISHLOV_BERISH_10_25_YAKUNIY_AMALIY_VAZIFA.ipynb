#**6) APIʼlar bilan ishlash**
1. Yuqorida berilgan kriptovalyuta maʼlumotlaridan `id` ustunini roʻyxat koʻrinishida saqlang va bu roʻyxatga `coin_ids` deb nom bering.
2. https://api.coingecko.com/api/v3/simple/price?ids={coin_ids}&vs_currencies=usd havolasi orqali kriptovalyutalarning `usd` ga nisbatini oling (birinchi 100 ta maʼlumotni olish yetarli).
3. 2-shartda berilgan link orqali olingan maʼlumotlarni dataframe koʻrinishida saqlang. Bunda uning indeksi va `usd` ustuni boʻlsin. `index` -> kriptovalyutalar nomi, `usd` -> narxi

#**💡Ishora:** linkdagi `{coin_ids}` oʻzgaruvchi sifatida yozilgan. Siz ham 1-shartda saqlagan `coin_ids` oʻzgaruvchingizni shu qismda yozib keting!


````python
import pandas as pd

df_coins_list = pd.read_json(coins_url)  # df_coins_list ga jadval saqlandi
coin_ids_full = df_coins_list["id"].tolist()
# tolist() yordamida id ustuni malumotlarini coin_ids_full o'zgaruvchisiga list qilindi .

coin_ids = coin_ids_full[:100]
display(f"birinchi 100 ta coin IDsi dan 5 slicing qilindi: {coin_ids[:5]}...")
print("\n")
# coin_ids_full dagiid ning 100 tasi coin_ids ro'yxatiga saqlandi va display bilan 5 tasi chiqarildi.

price_url = f"https://api.coingecko.com/api/v3/simple/price?ids={','.join(coin_ids)}&vs_currencies=usd"
# Bu yerda f-string yordamida coin_ids ro'yxatidagi barcha ID'lar vergul bilan ajratilib, URL ichiga kiritiladi.
# vs_currencies=usd esa narxlar AQSh dollari (USD) bo'yicha olinishini bildiradi.

df_prices = pd.read_json(price_url)
# Yuqorida yaratilgan price_url orqali API'ga so'rov yuboradi, olingan JSON javobni o'qiydi
# va uni df_prices nomli pandas DataFrame'iga aylantiradi. Bu DataFrame har bir tanlangan kriptovalyutaning USD'dagi narxini o'z ichiga oladi.

display(df_prices.head())
````

````python
#javobni shu yerda yozing.
# Example of an f-string query to search customers by FirstName and LastName
# Replace 'FirstName' and 'LastName' with actual values you want to search for

search_first_name = 'Luís'
search_last_name = 'Gonçalves'

query = f"""
SELECT * FROM customers
WHERE FirstName = '{search_first_name}' AND LastName = '{search_last_name}'
"""

# Execute the query and display results
df_search_results = pd.read_sql_query(query, db)
display(df_search_results)

# You can also search for parts of names using LIKE
search_term = 'L%'
query_like = f"""
SELECT * FROM customers
WHERE FirstName LIKE '{search_term}' OR LastName LIKE '{search_term}'
"""

df_like_results = pd.read_sql_query(query_like, db)
print(f"\nCustomers with FirstName or LastName starting with '{search_term}':")
display(df_like_results)
````
