# **HDF5 Formati**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import h5py
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv', index_col=0, names=['hudud','maydon','aholi'])
df1
```

```python
df1.to_excel("uzdata.xlsx") # Yuqoridagi jadvalni salab oldik
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv', index_col=0)
df2
```

```python
df2.to_excel("newcars.xlsx") # Yuqoridagi jadvalni salab oldik
```

```python
hdfobj = pd.HDFStore("data.h5")
# hdfobj degan o'zgaruvchi fayl yaratayabmiz,
# hamda pd.HDFStore degan class yordamida data.h5 degan fayl yaratdik
# Shuningdek hdfobj ga nimani saqlasak data.h5 fayliga saqlanadi
```

```python
hdfobj.put("uzdata",df1)
# Bunda hdfobj.put bilan df1 ni yani uzb viloyatlarini uzdata nomi bilan data.h5 ga saqladik
```

```python
hdfobj.put("cardata",df2)
# Bunda esa df2 degan mashina malumotlarini ham cardata degan nom bilan data.h5 ga saqladik
```

```python
hdfobj.keys()
# Bunda hdfobj ga saqlagan fayllarimiz nomi chiqadi.
# 2 ta yani cardata va uzdata degan faylni saqlagan edik. cars esa cardata ning
# eski saqlangan nomi bo'lgani uchun chiqardi
```
