3. Saqlab olingan `aqi.csv` faylining oxirgi 5 ta qatorini oʻqing.

Eslatma: fayl nomini `aqi.csv` deb saqlaganimizning sababi AQI – Air Quality Index, yaʼni havo sifati indeksi deganidir. Bizning maʼlumotlarimizda davlatlarning havo sifati haqidagi maʼlumotlar saqlangan!

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/data_date.csv"
#javobni shu yerda yozing.
aqi.csv = pd.read_csv("https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/data_date.csv")
aqi.csv.tail(5)
```

#2-topshiriq
1. Saqlab olingan `aqi.csv` faylidagi maʼlumotlarning faqat 2024-yilga tegishlisini excel faylga `aqi_2024.xlsx` koʻrinishida saqlab oling.
2. Saqlab olingan `aqi_2024.xlsx` faylidan Oʻzbekistonga tegishli maʼlumotni (qatorlarni) oʻqing.

```python
aqi['Date'] = pd.to_datetime(aqi['Date'])
aqi_2024 = aqi[aqi['Date'].dt.year == 2024]
aqi_2024.to_excel("aqi_2024.xlsx", index=False)

aqi_uzbekistan_2024 = pd.read_excel("aqi_2024.xlsx")
aqi_uzbekistan_2024 = aqi_uzbekistan_2024[aqi_uzbekistan_2024['Country'] == 'Uzbekistan']
display(aqi_uzbekistan_2024)
```

#3-topshiriq
1. `aqi.csv` faylidagi maʼlumotlarni oʻrganib chiqsangiz, bu faylda 2022–2025-yillar oraligʻidagi havo sifati haqidagi maʼlumotlarni koʻrishingiz mumkin. Hozir sizga vazifa:
  * `aqi_2022.csv`, `aqi_2023.csv`, `aqi_2024.csv`, `aqi_2025.csv`  fayllariga `aqi.csv` faylidagi maʼlumotlarning mos davrlarini saqlang. Misol uchun, `aqi_2022` fayli 2022-yildagi havo sifati haqidagi maʼlumotlarni oʻzida saqlaydi.
  

```python
for year in range(2022, 2026):
    # Har bir yil uchun ma'lumotlarni filtrlaymiz
    df_year = aqi[aqi['Date'].dt.year == year]

    # Fayl nomini yaratamiz (masalan, aqi_2022.csv)
    file_name = f"aqi_{year}.csv"

    # Filtrlagan ma'lumotlarni CSV fayliga saqlaymiz
    df_year.to_csv(file_name, index=False)
```

```python
import pandas as pd

aqi_2022_df = pd.read_csv('aqi_2022.csv')
display(aqi_2022_df.head())

aqi_2023_df = pd.read_csv('aqi_2023.csv')
display(aqi_2023_df.head())

aqi_2024_df = pd.read_csv('aqi_2024.csv')
display(aqi_2024_df.head())

aqi_2025_df = pd.read_csv('aqi_2025.csv')
display(aqi_2025_df.head())
```

#4-topshiriq
1. Saqlab olingan `aqi_2022.csv`, `aqi_2023.csv`, `aqi_2024.csv`, `aqi_2025.csv`  fayllarini HDF5 formatida saqlang. Fayl nomini `aqi.h5` deb nomlang. Kalit (key) sifatida fayllarning yillarini bering. Misol uchun, key=2025.

```python
#javobni shu yerda yozing.
aqi_h5 = pd.HDFStore("data.h5")

aqi_h5.put("2022",aqi_2022_df)
aqi_h5.put("2023",aqi_2023_df)
aqi_h5.put("2024",aqi_2024_df)
aqi_h5.put("2025",aqi_2025_df)

a = aqi_h5.keys()
print(f"\n Bu data.h5 ga saqlagan fayllarni ko'rsatadi => \n\n{a}\n")

aqi_h5["2022"].head()
```

#5-topshiriq
1. `aqi.h5` faylidan maʼlumotlarni oʻqing va 2025-yilga tegishli maʼlumotlarni yangi dataframeʼga saqlang.
2. `aqi.h5` faylida 2025-yilning qaysi oylari uchun maʼlumotlar bor? Shu oylar roʻyxatini ekranga chiqaring!

```python
# 1. `aqi.h5` faylidan maʼlumotlarni oʻqing va 2025-yilga tegishli maʼlumotlarni yangi dataframeʼga saqlang.
aqi_h5_store = pd.HDFStore("data.h5")
df_2025 = aqi_h5_store["2025"]

# 'Date' ustunini datetime formatiga o'zgartirish (agar hali o'zgarmagan bo'lsa)
df_2025['Date'] = pd.to_datetime(df_2025['Date'])

# 2. `aqi.h5` faylida 2025-yilning qaysi oylari uchun maʼlumotlar bor? Shu oylar roʻyxatini ekranga chiqaring!
months_2025 = df_2025['Date'].dt.month.unique()

print("2025-yil uchun ma'lumotlar mavjud bo'lgan oylar ro'yxati (raqamlar bilan):")
print(sorted(months_2025.tolist()))

month_names = {1: 'Yanvar', 2: 'Fevral', 3: 'Mart', 4: 'Aprel', 5: 'May', 6: 'Iyun',
               7: 'Iyul', 8: 'Avgust', 9: 'Sentabr', 10: 'Oktabr', 11: 'Noyabr', 12: 'Dekabr'}
named_months_2025 = [month_names[m] for m in sorted(months_2025.tolist())]
print("2025-yil uchun ma'lumotlar mavjud bo'lgan oylar ro'yxati (nomlari bilan):")
print(named_months_2025)

display(df_2025.head())
```
