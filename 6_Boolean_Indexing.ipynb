# **2.1**.1. N-o'lchovli massiv (array)larga ishlov berish
# Bolean indekslash

```python
# ismlar
names = np.array(['Hasan', 'Husan', 'Javohir', 'Elyor', 'Hasan', 'Javohir', 'Elyor'])
names
```

```python
# data
data = np.random.randn(7, 4) # Bu 7 qator va har qatorda 4 tadan malumot yaratadi
data
```

```python
names == 'Hasan' # Ismlarda 2 ta hasan bo'lgani uchun 2 ta True bor
```

```python
data[names == 'Hasan'] # Hasanlarga tegishli barcha malumotlarni ajratib oladi
```

```python
data[names == 'Elyor']
```

```python
data[names == 'Husan'] # Bunda husan ga tegishli malumotlar chiqadi
```

```python
data[names == 'Husan', 2:] # Husan ga tegishli oxirgi 2 ta malumotni chiqarib olish
```

```python
data[names != 'Hasan'] # Hasandan tashqari boshqa barcha odamlarning malumotini chiqarish
```

```python
data[~(names == 'Hasan')] # Bu ham yuqoridagi buyruq bilan bir xil vazifani bajaradi
```

```python
print(names)
print(data)
```

```python
ikkiism = (names == 'Hasan') | (names == 'Husan') # "|" - or (yoki) degan manoni anglatadi
ikkiism # Bunda hasan va husan bor joylar True bilan chiqadi
```

```python
data[ikkiism] # Hasan va Husan ga tegishli malumotlarni ajratib oladi
```

```python
data[data<0] # Bunda 0 dan kichik malumotlarni chiqaradi
```

```python
data[data<0] = 0 # Bunda 0 dan kichik sonlarni 0 ga tenglab qaytadan malumotlarni chiqaramiz
data
```
