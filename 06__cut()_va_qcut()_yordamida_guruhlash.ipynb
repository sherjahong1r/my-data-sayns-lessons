```python
import pandas as pd
import numpy as np
```

## **Quyidagi oraliqda nechtadan davlat borligini tahlil qilamiz:**

1 milliongacha (0-1000)

1 - 10 mln oraliqda (1000-10000)

10 - 30 mln oraliqda (10000-30000)

30 - 50 mln oraliqda (30000-50000)

50 - 100 mln oraliqda (50000-100000)

100 - 300 mln oraliqda (100000-300000)

300 - 1.5 mlrd oraliqda (300000-1_500_000_000)


**Buning uchun quyidagi bin (nuqtalarni) yaratib olamiz:**


```python
bins = [0, 1000, 10000, 30000, 50000, 100000,  300000, 1500000]
# 7 ta oraliqqa ajratib oldik
```

```python
population = df.pop2021
```

```python
population
# Aholini ko'rsatgan ustunni ajratib oldik yani pop2021 ni.
```

```python
guruh = pd.cut(population, bins)
guruh
# .cut yordamida population va bins o'zgaruvchilarini qo'ysak, bins dagi qaysi oraliq guruhlarga kirishini ko'rsatadi.
```

```python
guruh.value_counts()
# Qaysi aholi guruhlari orasida nechta davlatlar soni mavjud ekanligini ko'rsatadi
# Biz quyida 7 ta aholi oraliq guruhlarini yaratganmiz.
```

```python
df['guruh'] = guruh
df
# guruh ustunini asl df jadvaliga qo'shdik
```

```python
df2 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df2
```

```python
df2.price.describe()
# Bu bilan min, max, std, mean va shu kabi boshqa ma'lumotlarini bilishimiz mumkin.
```

```python
bins2 = [0, 10000, 20000, 50000]
# 0 dan 10_000 gacha,
# 10_000 dan 20_000 gacha,
# 20_000 dan 50_000 gacha bo'lganlarni 3 ta guruhga ajratdik
```

```python
range = pd.cut(df2.price, bins2)
range
# Bunda df2 jadvalini price ustunini guruhlarga ajratdik
```

```python
range.value_counts()
# guruhlardagi davlatlar soni.
```

```python
names = ["arzon", "o'rta", "qimmat"]
range = pd.cut(df2.price, bins2, labels=names)
range.value_counts()
# labels dan foydalanib 3 ta guruhga bo'lgandagi mashinalar sonini ko'rishimiz mumkin
```

```python
df2["arzon_o'rta_qimmat"] = range
df2
# yuqoridagi ustunni df2 jadvaliga qo'shdik.
```

```python
range = pd.cut(df2.price, 4)
range.value_counts()
# Yuqorida 3 guruhga bo'lib o'zgaruvchiga saqlagan edik
# Bunda shunchaki nechta guruhga bo'lmoqchi bo'lsak raqamini berib qo'ysak bo'ldi
```

```python
range = pd.qcut(df2.price, 4)
range.value_counts()
# qcut yordamida nisbatan mashinalar sonini teng taqsimlagan holda bo'lishimiz mumkin yuqoridagidan farqli ravishda.
```

```python
df2['new_price'] = range
df2
# 3 ta guruhga bo'lgandagi ustunni asl jadvalga qo'shdik.
```

```python
range1 = pd.qcut(df2.horsepower, 3)
range1.value_counts()
# Bunda boshqa yani horsepower ustunini guruhladik qcut bilan.
```

```python
df2['new_horsepower_2'] = range1
df2
```
