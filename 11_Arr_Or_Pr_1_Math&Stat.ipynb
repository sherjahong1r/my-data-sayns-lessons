```python
# data
arr = np.random.randn(5, 4) # (5, 4) o'lchamli massiv (random)
arr
```

## sum metodi

```python
arr.sum()
```

```python
print(arr.sum()) # arr ning barcha elementlarini yig;indisini hisoblab beradi
```

```python
np.sum(arr) # Bu ham shu hisobni bajaradi
```

```python
# Bunda pastda ustun va qatorga hisoblash amalga oshirilga index bilan. Index gorizantalga 1 dan, vertikalga 0 dan boshlanadi.
```

```python
arr.sum(axis=1) # gorizantal qilib yig'indini hisoblaydi
```

```python
arr.sum(axis=0) # ustun yani vertikal qilib yig'indini hisoblaydi
```

## mean metodi

```python
# sodda ma'lumot
data = np.arange(5) # 0 ~ 5 gacha bo'lgan elementlardan tashkil topgan massiv
data
```

```python
mean = np.sum(data)/len(data) # Bunda sum bilan hammasini qo'shib, len orqali nechtaligini aniqlab bo'layabmiz
print(mean)
```

```python
# mean metodi orqali o'rtacha qiymat aniqlanadi. sonlar qo'shilib nechta son bo'lsa shunga bo'linadi
# numpy mean
np.mean(data)
```

```python
data.mean() # Bu ham shu vazifani bajaradi
```

```python
arr
```

```python
np.mean(arr)
```

```python
ortacha = arr.mean()
print(ortacha) # arr ning o'rtacha qiymati
```

```python
np.mean(arr, axis=1) # ustunlar o'rtacha qiymati yani gorizantal bo'yicha
```

```python
np.mean(arr, axis=0) # qatorlar o'rtacha qiymati yani vertikal bo'yicha
```
