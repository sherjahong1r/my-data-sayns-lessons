# N-o'lchovli massivlar yaratishning boshqa **usullari**

zeros va ones funksiyalari yordamida massivlar yaratish

```python
arr5 = np.zeros((2, 4)) # barcha elementlar 0 ga teng (2, 4)
arr5
```

```python
arr6 = np.ones((2, 4)) # barcha elementlar 1 ga teng (2, 4)
arr6
```

arrange funksiyasi yordamida yaratish

```python
arr7 = np.arange(4, 20, 2) # 4 dan 20 gacha sonlarni 2 qadamlab chiqaradi
arr7
```

```python
arr8 = np.arange(20) # 0 dan 20 gacha 1 qadamlab chiqaradi
arr8
```

```python
import numpy as np

# empty funksiyasi - o'rnatilgan qiymatlarsiz, lekin ko'rsatilgan shaklga ega massiv yaratadi.
# Dastlabki qiymatlar xotirada mavjud bo'lgan narsalar bo'ladi (tasodifiy raqamlar emas).
arr_empty = np.empty((2, 3))
print("np.empty((2, 3)):\n", arr_empty)

# full funksiyasi - berilgan qiymat bilan to'ldirilgan massiv yaratadi.
arr_full = np.full((3, 2), 7) # 3 qator, 2 ustun, barcha elementlar 7
print("\nnp.full((3, 2), 7):\n", arr_full)

# eye funksiyasi - diagonalda birlar va boshqa joylarda nollar bo'lgan ikki o'lchovli massiv (identifikatsiya matritsasi) yaratadi.
arr_eye = np.eye(4) # 4x4 identifikatsiya matritsasi
print("\nnp.eye(4):\n", arr_eye)

# identity funksiyasi - eye bilan bir xil, kvadrat identifikatsiya matritsasi yaratadi.
arr_identity = np.identity(3) # 3x3 identifikatsiya matritsasi
print("\nnp.identity(3):\n", arr_identity)

# linspace funksiyasi - berilgan oraliqda bir tekis joylashgan raqamlar ketma-ketligini yaratadi.
arr_linspace = np.linspace(0, 10, 5) # 0 dan 10 gacha 5 ta son
print("\nnp.linspace(0, 10, 5):\n", arr_linspace)

# random.rand funksiyasi - 0 dan 1 gacha bo'lgan tasodifiy sonlar massivini yaratadi.
arr_rand = np.random.rand(2, 2) # 2x2 tasodifiy sonlar massivi
print("\nnp.random.rand(2, 2):\n", arr_rand)
```
