```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head() # Yuqoridagi fayl silkasini .read_csv bilan ochyabmiz hamda boshidagi beshtasini .head() bilan chiqarayabmiz
```

```python
df.shape # Buning hajmi juda katta yani 2499 qatorga 12 ustun. AQSH mashinalarining sotilayotgan jadvali
```

```python
df.sort_index() # Bu avvaldan tartiblangani uchun o'zgarmadi
```

```python
df.sort_index(axis=1, inplace=True) # Bu ustunlarni tartiblaydi Alifbe tartibida. Hamda inplace bilan saqlab qolyabdi
df
```

```python
df.sort_index(ascending=False) # Bu teskari tartibda tahlaydi yani kattadan kichikga
```

```python
df.sort_index(axis=1, ascending=False) # Bunda ustunlarni teskari tartibda joylashtiradi
```

```python
df.sort_values(by='price') # Bu qiymatlar bo'yicha tartiblaydi yani price ning kichikdan kattaga o'sadi
```

```python
df.sort_values(by='price', ascending=False) # Bu teskari tartibda tartiblaydi yani katta narxdan kichik narxga kamayadi
```

```python
df.sort_values(by=['price','mileage'], ascending=False) # Bunda price va mileageni kattadan kichikga tartiblaydi
```

```python
df.sort_values(by='year')
```
