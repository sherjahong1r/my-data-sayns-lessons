# **Tarqoqlik grafigi (scatterplot)**

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="year", y="price")
plt.show()
# Bu grafikka scatterplot grafigi deyiladi
# Bu yerda 13119 ta nuqta bor degani chunki yuqoridagi jadvalda shuncha qator bor
# Hamda ahamiyat bersak mashinalar yili oshgan sari ularning narxi ham oshib boryabdi
# Va 1970 yilga qarasak 1 ta dona mashina bor ekan bu bizning grafikga salbiy ta'sir ko'rsatadi shu sababli uni olib tashlayamiz
```

```python
df = df[df.year>2000]     # df = df[.year!=1970]  yokida bunday teng emas degan metoddan ham foydalansak bo'lardi
# bunda data ga 2000 yildan katta bo'lgan mashinlarni uzatayabmiz
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="year", y="price")
plt.show()

# bunday grafikga pozitiv korrelyatsiya deymiz chunki yili oshgani sayin uning narxi ham oshib borayabdi
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
#sns.scatterplot(data=df, x="year", y="price")
sns.regplot(data=df, x="year", y="price")
# regplot bilan o'suvchi korrelyatsiya bo'layotganini tekshirish mumkin
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="mileage", y="price")
#sns.regplot(data=df, x="mileage", y="price")
plt.show()
# Bunga etibor bersak ko'p yurgani sayin uning narxi pasayib boryabdi
# Bunga negativ korrelyasiya deyiladi
```

```python
#df = df[df.year>2000]
plt.figure(figsize=(15,8))
#sns.scatterplot(data=df, x="mileage", y="price")
sns.regplot(data=df, x="mileage", y="price")
plt.show()
# Bunga etibor bersak ko'p yurgani sayin uning narxi pasayib boryabdi
# Bunga negativ korrelyasiya deyiladi
```

## Keling grafigimizga 3-o'zgaruvchi, mashina korobkasi turini ham qo'shamiz (transmission) buning uchun hue parametridan foydalanamiz.

```python
#df = df[df.year>2000]
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x="mileage", y="price", hue='transmission', s=100)
#sns.regplot(data=df, x="mileage", y="price")
plt.show()
# hue='transmission'   bu bilan ranglarga ajratamiz
# s=100  bu nuqtalarning hajmi yani kattaligi
# mashinalarning karobkasi - automatic - ko'k, Manual - sariq, Semi-Auto - yashil, Other - qizil bilan belgilangan
```

##Grafikdan qanday xulosalar qilishimiz mumkin:

Aksar mashinalar avtomat korobka (Automatic)
Mexanik (Manual) korobka mashinalar juda arzon (sariq rang)
Yarim-avto (Semi-auto) mashinalar nisbatan qimmat

```python
df.head(1)
```

```python
df['fuelType'].unique()
# Demak FuelType ustinida 4 ta takrorlanmas qiymatlar mavjud ekan.
```

```python
#df = df[df.year>2000]
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x="mileage", y="price", hue='transmission', style='fuelType', s=100)
#sns.regplot(data=df, x="mileage", y="price")
plt.show()
# style='fuelType' bu fueltype ustuni ham qo'shildi shartli belgisida ko'rsatilgan qaysi yoqilg'ida yurishi shakllar bilan
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x="mileage", y="price", hue='fuelType', s=100)
plt.show()
# Bunda faqat hue bilan fuelTypeni yani qayasi yoqilg'ida yurishini tekshirdik
```

Scatter plotning ayan bir ko'rinishi bu Swarm plot. swarmplot yordamida yuqoridagi har bir kategoriyani (transmission) alohida swarm (to'da) ko'rinishida chiqarish mumkin.

Diqqat! swarmplotda kategoriyalar soni qancha ko'p bo'lsa, funksiya shuncha uzoq vaqt ishlaydi.

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
# swarmplot - Nuqtalarni tarqatish: U ma'lumotlar nuqtalarini bir-birining ustiga chiqmasdan,
# ularning taqsimotini aniq ko'rsatish uchun "suruv" shaklida joylashtiradi. Bu, ayniqsa,
# bir kategoriya bo'yicha ko'plab nuqtalar mavjud bo'lganda foydalidir,
# chunki odatiy scatterplot da nuqtalar bir-birini yashirib qo'yishi mumkin.
```

Yuqoridagi jadvaldan ham turli xulosalar qilish mumkin:

Automatic va Semi-auto mashinalar deyarli bir hil narx kategoriyada
Manual mashinalar arzon mashinalar segmentida (30000$ gacha).
scatterplot haqida ko'proq: https://seaborn.pydata.org/generated/seaborn.scatterplot.html
