# **12**-**Dars**
# Tartiblash (ORDER BY)
# Saralash (DISTINCT)



```python
komanda = "SELECT CustomerId, Firstname, LastName, Phone, Country FROM customers"
df = pd.read_sql_query(komanda, db, index_col='CustomerId')
df.head()
# Chiqarmoqchi bo'lgan ustunlarimizni komanda o'zgaruvchisiga saqlab df ga komanda deb yozib chiqarildi.
```

```python
komanda = "SELECT CustomerId, Firstname, LastName, Phone, Country FROM customers ORDER BY FirstName LIMIT 20"
df = pd.read_sql_query(komanda, db, index_col='CustomerId')
df
# ORDER BY bilan FirstName ustunini alifbe tartibida joylashtirayabdi.
# LIMIT 20 bilan faqat 20 qatorini chiqarayabdi
```

```python
query = "SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
# DISTINCT bilan takrorlanmas Country ustunidagi davlatlar chiqarildi
# fetchall dan foydalanganimiz uchun ro'yxat ko'rinishida chiqardi.
```

```python
pd.read_sql(query, db)
# pandas DataFrameda jadval ko'rinishida chiqarildi
```

# **13-dars**
# Filtrlash (WHERE)

```python
query = 'SELECT * FROM customers WHERE country="Italy"'
pd.read_sql(query, db)
# Bu cod - WHERE yani mamlakati Italiyaga teng bo'lgan barcha ro'yxatni chiqar degani
# Shuningdek biz codlarni hammasini katta harflard yozganimi kichik harflaeda ham yozsa bo'ladi
# Faqat codlar ajralib turilishi uchun yozilgan.
```

```python
query = 'select * from customers WHERE country="USA"'
pd.read_sql(query, db)
# USA da ko'proq yani 13 qator bor ekan.
```

```python
query = 'SELECT * FROM customers WHERE country="Brazil" OR country="Argentina" ORDER BY FirstName'
pd.read_sql(query, db)
# OR bilan Brazil va Argentina davlatlarini birdaniga chiqarayabmiz
# ORDER BY bilan FirstName ustunini alifbe tartibida ham taxlandi. Hamda buni Countryga nisbatan ham qo'llashimiz mumkin edi.
```

```python
query = 'SELECT * FROM customers WHERE country="USA" AND State="CA" ORDER BY Country'
pd.read_sql(query, db)
# Yuqoridagilarda OR bilan yo unisi yo bu shart bajarilsa ham chiqarardi
# AND bilan 2 ta shart ham qanoatlanishi kerak yani faqat Countrysi USA ga teng va State si CA ga teng bo'lganlar chiqadi
```

```python
query = 'SELECT * FROM customers WHERE country="USA" OR State="CA" ORDER BY Country'
pd.read_sql(query, db)
# OR da yo unisi yo bunisi bo'lganlar ham chiqqani uchun qatorlar soni ko'paydi
```

```python
query = "SELECT * FROM customers ORDER BY Country DESC"
df = pd.read_sql(query, db, index_col="CustomerId")
df.head(10)
```

```python
# Brazil va USA xaridorlarni ajratib olamiz
# xaridorlar sonini chiqaramiz
cursor.execute('SELECT COUNT (*) FROM customers WHERE country="Brazil" OR country="USA"')
print(cursor.fetchone())
```

```python
# Brazil va USA xaridorlarni ajratib olamiz
query = 'SELECT Firstname, LastName, Country, Phone FROM customers WHERE country="Brazil" OR country="USA" ORDER BY Country'
pd.read_sql(query, db)
```
