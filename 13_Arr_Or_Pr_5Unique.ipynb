## Unique va inld setdiffld va boshqa usullari haqida

```python
# data
names = np.array(['Sarvar', 'Abdurahmon', 'Hasan', 'Temur', 'Sarvar', 'Temur'])
ints = np.array([3, 3, 3, 2, 2, 1, 1, 4, 4])
```

```python
# unique hamda alifbe tartibida ham joylashtiradi
np.unique(names) # names dagi takrorlanmas elementlarni qaytaradi (takrorlansa ham faqat bittasini qabul qiladi)
```

```python
np.unique(ints)
```

```python
set(names) # names dagi unique(takrorlanmas) elementlarni qaytaradi
```

## inld usuli haqida

```python
# data
arr1 = np.array([6, 0, 0, 3, 2, 5, 6]) # Bu arr2 elementlari arr1 da bor bo'lsa True qaytaradi
arr2 = np.array([0, 2, 3])
```

```python
np.isin(arr1, arr2) # inld deganimda xato chiqdi shuning uchun insin deb yozdim va to'g'ri chiqdi
```

## setdiffid usuli haqida

```python
# data Buning ham asli yozilishi setdiff1d ekan
names1 = np.array(['Jasur', 'Abdurahmon', 'Hasan', 'Muhammad', 'Sarvar', 'Temur'])
names2 = np.array(['Sarvar', 'Abdurahmon', 'Hasan', 'Temur', 'Sarvar', 'Temur'])
```

```python
np.setdiff1d(names1, names2) # names1 massivining names2 massividagi takrorlanmas qismini qaytaradi
```
