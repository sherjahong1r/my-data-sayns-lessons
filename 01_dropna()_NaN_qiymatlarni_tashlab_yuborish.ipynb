```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df
```

```python
df.isnull()
# Qiymat mavjud bo'lmagan yani NaN lar o'rniga True qaytaradi, aksi bo'lsa False
```

```python
df.notnull()
# Bu yuqoridagining aksi yani qiymat mavjud bo'lsa True, yo'q yani NaN bo'lsa false qaytaradi.
```

```python
df.isnull().sum()
# Bu har bir ustundagi NaN bo'lganlar sonini ko'rsatadi
```

```python
df.notnull().sum()
# Ustunlardagi mavjud qiymatlar sonini ko'rsatadi
```

```python
len(df)
# asl df jadvali 61 qator edi
```

```python
df.dropna(axis=1)
# Bunda NaN qiymati mavjud bo'lgan ustunlarni tashlab yubordik
```

```python
len(df.columns)
# umumiy ustunlar soni
```

```python
len(acd)
# Umumiy 61 qator mavjud edi va price ustunidagi NaN qatorlar o'chirilgandan keyin 58 qator qolibdi
# Shunda price ustunida 3 qatorida NaN mavjud bo'lgan
```
