#3) Har bir ustundagi boʻsh qiymatlarni aniqlang va ularni toʻldirish strategiyasini ishlab chiqing.

```python

df.isnull()
```

```python
df.isnull().sum()
```

```python
# 'Category' ustunidagi bo'sh qiymatlarni eng ko'p uchraydigan qiymat (mode) bilan to'ldirish
mode_category = df['Category'].mode()[0]
df['Category'].fillna(mode_category, inplace=True)

# Tekshirish
print(df['Category'].isnull().sum())

# .mode(): Bu usul tanlangan ustundagi (bizning holatda Category ustunidagi) eng ko'p uchraydigan qiymatni yoki qiymatlarni qaytaradi.
#  Agar bir nechta qiymat bir xil chastotada eng ko'p uchrasa, .mode() ularning barchasini Series sifatida qaytaradi.

# [0]: .mode() usuli bir yoki bir nechta eng ko'p uchraydigan qiymatlarni qaytarishi mumkin bo'lgani uchun,
#  u natijani Series obyektida saqlaydi. [0] indeksidan foydalanish, agar bir nechta moda qiymatlari bo'lsa,
#  ulardan faqat birinchisini (ya'ni, Seriesdagi birinchi elementni) olishni ta'minlaydi.
```

```python
# 'Item' ustunidagi bo'sh qiymatlarni 'Noma'lum' bilan to'ldirish
df['Item'].fillna('Others', inplace=True)
print(df['Item'].isnull().sum())
```

```python
# 'Quantity' ustunidagi bo'sh qiymatlarni median bilan to'ldirish
median_quantity = df['Quantity'].median()
df['Quantity'].fillna(median_quantity, inplace=True)

# Barcha bo'sh qiymatlar to'ldirilganini tekshirish
print(df['Quantity'].isnull().sum())

# [1, 2, 3, 4, 5] ma'lumotlar to'plamining mediani 3.
# [1, 2, 3, 4, 5, 6] ma'lumotlar to'plamining mediani (3+4)/2 = 3.5.
# median bilan shunda o'rtadagi qiymatni olib NaN lar o'rniga qo'ydik
```

```python

np.mean(df['Order Total'])
```

```python
df.fillna({"Order Total":np.mean(df['Order Total'])}, inplace=True)
# Order Total ustunini o'zining o'rtacha qiymatlari bilan to'ldirdik NaN lar o'rnini
```

```python
# df.fillna(method='ffill')
# NaN bo'lgan qatorga o'zidan bitta oldingi qatordagi qiymatni qo'yadi.

df['Payment Method'].fillna(method='bfill', inplace=True)
print(df['Payment Method'].isnull().sum()) # Tekshirish uchun
# Bitta pastdagi qiymat bilan to'ldirildi
```

#8) Mahsulot nomlarida noto‘g‘ri yoki nomaʼlum belgilar (?, NaN, #) mavjudligini tekshiring va ularni tuzating.

```python
#javobni shu yerda yozing
# df['Item'].unique()
df['Item'] = df['Item'].str.replace('[^\w\s]', '', regex=True)
df['Item'].unique()

# ^: Bu belgilar to'plamining ichida bo'lsa, inkor ma'nosini bildiradi. Ya'ni, "quyidagilardan tashqari har qanday belgi".
# \w: Bu alifbo harflari (katta va kichik), raqamlar (0-9) va pastki chiziqni (_) o'z ichiga olgan belgilarni anglatadi.
# Umumiy qilib aytganda, "word characters" (so'z belgilari)
# \s: Bu bo'sh joy belgilari (space, tab, newline va boshqalar) ni anglatadi.
# Demak, [^\w\s] "so'z belgilaridan va bo'shliqlardan tashqari har qanday belgi" degan ma'noni beradi.
# '' (bo'sh satr): Bu muntazam ifoda topgan belgilarni nima bilan almashtirish kerakligini bildiradi.
# Bo'sh satr bilan almashtirish, topilgan maxsus belgilarni o'chirishni anglatadi.
# regex=True: Bu replace funksiyasiga birinchi argument muntazam ifoda ekanligini bildiradi. Bu uni oddiy matn almashtirishdan ajratib turadi.
# df['Item'].unique() bilan bajarilgan amallarni chiqaradi.
```

## Correcting negative values in 'Price' column

### Subtask:
Correct any negative values in the 'Price' column by taking their absolute value. This ensures all prices are positive, which is consistent with real-world sales data.

#### Instructions:
1. Apply the absolute value function (`abs()`) to the 'Price' column to convert all negative values to their positive counterparts. Assign the result back to the 'Price' column.

**Reasoning**:
