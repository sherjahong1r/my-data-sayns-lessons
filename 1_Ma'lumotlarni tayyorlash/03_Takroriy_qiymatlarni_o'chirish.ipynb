# **Takroriy qiymatlarni o'chirish**.
# **Qiymatlarni o'zgartirish**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv')
df.head()
```

```python
df.duplicated()
# Bu takrorlangan qiymatlarni ko'rsatadi, takrorlangan bo'lsa True bo'ladi, aks holda False qaytaradi.
```

```python
df.shape
# qatorlar sonini aniqlandi.
```

```python
df.duplicated().sum()
# duplicated bo'lgan yani qiymatlari takrorlangan qatorlar soni aniqlandi.
```

```python
df.drop_duplicates()
# dublicated bo'lgan qatorlarni tashlab yubordik drop bilan va endi asl jadvaldagi 247 ta qatordan, 232 ta qator qoldi
# Etibor beraylik duplicates deb yozildi yuqoridagilarda duplicated yozilayotgan edi.
```

```python
df.drop_duplicates(['country','rank'])
# Bu kod df DataFrame'idan 'country' va 'rank' bo'yicha takrorlangan qatorlarni olib tashlab, noyoblarini qoldiradi.
```
