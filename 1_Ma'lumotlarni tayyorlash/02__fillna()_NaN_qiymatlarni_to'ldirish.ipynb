# **Ma'lumotlarni to'ldirish**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.fillna(0)
# NaN bo'lgan qiymatlar o'rniga 0 qo'yiladi
```

```python
df.fillna({'horsepower':100, "wheel-base":90})
# Biz tanlagan ustunlardan horsepower ga NaN lar o'rniga 100, wheel-base ga 90 qo'yiladi.
```

```python
np.mean(df['wheel-base'])
# wheel-base ning o'rtacha qiymati hisoblandi
```

```python
np.mean(df['horsepower'])
```

```python
df.fillna({'horsepower':np.mean(df['horsepower']), "wheel-base":np.mean(df['wheel-base'])})
# NaN lar o'rniga tegishli ustunning o'rtacha qiymati qo'yildi.
```

```python
df
```

```python
df.fillna(method='ffill')
# NaN bo'lgan qatorga o'zidan bitta oldingi qatordagi qiymatni qo'yadi.
```

```python
df.fillna(method='bfill')
# Bunda NaN o'rniga o'zidan bitta pas qatordagi qiymat qo'yiladi.
```

```python
# df.fillna(method='ffill'): Bu usul "forward fill" degani bo'lib,
# o'zidan oldingi mavjud qiymatni keyingi bo'sh (NaN) kataklarga to'ldiradi.

# df.fillna(method='bfill'): Bu usul "backward fill" degani bo'lib,
# o'zidan keyingi mavjud qiymatni oldingi bo'sh (NaN) kataklarga to'ldiradi.
```

```python
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
# companiyasi toyotaga teng bo'lgan qatorlarda bfill filtrlash amalga oshirildi
```

```python
# yuqoridagi o'zgarishlar asl jadvalga saqlanmayotgan edi

df[df.company=='toyota'] = df[df.company=='toyota'].fillna(method='bfill')

# Bu metod bilan saqlanadi yani = belgisi bilan
```

```python
df[df.company=='toyota']
# Yuqoridagi o'zgarish saqlandi
```

```python
df
```

```python
# df.fillna(df.mean())
# Bu metod ham to'g'ri bo'lar edi agar hamma ustunlar sonlardan tashkil topganda
# Shuning uchun pastdagi metod bilan faqat sonli ustunlar uchun bajarayabmiz

numeric_cols = df.select_dtypes(include=np.number).columns # faqat sonli ustunlar numeric_cols ga saqlab olindi
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean()) # har bir sonli ustunlarning o'rtachasi hisoblanib NaN lar o'rniga qo'yildi.
display(df) # saqlangan o'zgaruvchilar display bilan chiqarildi.
```
