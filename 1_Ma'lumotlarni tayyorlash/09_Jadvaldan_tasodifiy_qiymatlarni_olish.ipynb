# **Jadvaldan tasodifiy qiymatlarni olish**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv", index_col=0)
df
```

```python
len(df)
# Demak 100 qator tartib sonli yani kichikdan kattaga o'sgan jadval mavjud
```

```python
randlist = np.random.permutation(100)
randlist
# Bu yordamida biz bu tartib raqamlarni aralashtirayabmiz
```

```python
df.take(randlist)
# take bilan yuqorda aralashtirgan id raqamlarini df jadvaliga qo'lladik
```

```python
df.sample(12)
# tasodifiy 12 qatorni ajratib oldik
```

```python
df.sample(12,replace=True)
# replace=True bilan esa tasodifiy 12 qator ichida bazi qatorlar yana takrorlangan bo'lishi mumkin
```
