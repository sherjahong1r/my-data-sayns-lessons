# **Arfimetik Amallar**

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
s1
```

```python
s2
```

```python
s1 + s2
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.add(df2)
```

```python
df1.add(df2, fill_value=10) # Bunda qiymat ikkisida umuman yoq bo'lsa nan bo'lib qoladi. Birida bor ikkinchisida yo'q bo'lsa fill_value ga bergan qiymatni qo'shib qo'yadi
```

```python
df2.sub(df1, fill_value=0)
```

```python
df2-df1
```

```python
df1.sub(df2, fill_value=0) # df1 dan df2 ayrilayabdi
```

```python
df2.rsub(df1, fill_value=0) # Bunda df2 birinchi kelishiga qaramasdan rsub bilan teskari yani df1 dan df2 ayrilayabdi
```

```python
df2.div(df1)
```

```python
df1.rdiv(df2) # Bu ham rdiv bilan yana df2 ni df1 ga bo'ladi
```

```python
df1.div(df2)
```

## Df va Series o'rtasida arfimetik amallar

```python
df = pd.DataFrame(np.arange(1,13.).reshape((3, 4)), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
obj = pd.Series([1,1,2,2], index=df.columns)
```

```python
df
```

```python
obj
```

```python
df+obj # Index va columns lar mos tushgani uchun df mevalari qiymatiga, obj mevalari qiymatini yani 1,1,2,2 larni qatorma qator qo'shayabdi
```
