## NumPyni chaqirib olamiz

```python
import numpy as np
```

```python
arr = np.arange(10)
```

```python
arr
```

```python
# Bu bilan fayl qilib saqlayabmiz
np.save('data', arr)
```

```python
# load bilan shu faylni o'qish mumkin
arr2 = np.load('data.npy')
```

```python
arr2 # load bilan arr2 degan o'zgaruvchida o'qiyabmiz
```

```python
arr1 = np.arange(20) # yangi arr1 massivi
```

```python
np.savez('arrays', a=arr, b=arr1) # savez bilan bitta fayl ichida 2 ta arr va arr1 degan massivni saqlayabmiz
```

```python
arrays = np.load('arrays.npz')
```

```python
arrays['a'] # alohida qilib arr ni o'zini o'qiyabmiz
```

```python
arrays['b'] # Bunda arr1 ni o'qiyabmiz
```

```python
np.savez_compressed('array_comp.npz', a=arr, b=arr1) # savez_compressed bilan faylni zip qilib
# yani siqib hajmini kichraytirib saqlashimiz mumkin
```
