### NumPy ma'lumot turlari (dtypes)

```python
import numpy as np

# Integer turlari
arr_int8 = np.array([1, 2, 3], dtype=np.int8)
print(f"int8: {arr_int8.dtype}")

arr_int16 = np.array([1, 2, 3], dtype=np.int16)
print(f"int16: {arr_int16.dtype}")

arr_int32 = np.array([1, 2, 3], dtype=np.int32)
print(f"int32: {arr_int32.dtype}")

arr_int64 = np.array([1, 2, 3], dtype=np.int64)
print(f"int64: {arr_int64.dtype}")

# Float turlari
arr_float32 = np.array([1.0, 2.0, 3.0], dtype=np.float32)
print(f"float32: {arr_float32.dtype}")

arr_float64 = np.array([1.0, 2.0, 3.0], dtype=np.float64)
print(f"float64: {arr_float64.dtype}")

# Boolean turi
arr_bool = np.array([True, False, True], dtype=np.bool_)
print(f"bool: {arr_bool.dtype}")

# String turi
arr_str = np.array(['hello', 'world'], dtype=np.str_)
print(f"string: {arr_str.dtype}")

# Complex turi
arr_complex = np.array([1+2j, 3+4j], dtype=np.complex64)
print(f"complex64: {arr_complex.dtype}")

arr_complex128 = np.array([1+2j, 3+4j], dtype=np.complex128)
print(f"complex128: {arr_complex128.dtype}")
```

Bu eng ko'p ishlatiladigan ma'lumot turlaridan ba'zilari. Har bir tur ma'lumotni saqlash uchun turli miqdordagi xotiradan foydalanadi va ularning chegaralari bor (masalan, `int8` kichikroq sonlarni saqlay oladi, `int64` esa juda katta sonlarni).
