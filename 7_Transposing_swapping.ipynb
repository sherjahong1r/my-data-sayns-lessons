# N-o'lchovli massiv (array)larga ishlov berish

## Massivlar pozitsiyasini ko'chirish transpose va o'qlar o'rnini almashtirish Swapping axes

```python
A = np.arange(6).reshape(3, 2) # A massive : Elementlari 0~6 gacha, shape (3, 2) yani reshape bilan vertikal 3 qator, gorizantal 2 qator degani
A
```

```python
# Transposening vazifasi gorizantalni vertikal qilib tahlaydi, va ko'pi bilan 2 massivlargacha bu amal ishlaydi
A.T #  Bu yerda T - Transpose degani yani massivning Transpose ko'rinishini beradi
```

## Swaping axes

```python
A
```

```python
A.swapaxes(1, 0) # A massivning Transpose ko'rinishi beradi, Bu usul bilan hohlagancha massivni transpose qila olamiz
```

## Transposening ishlatilinishiga bir misol AA,T Bunda T AA ning darajasida
