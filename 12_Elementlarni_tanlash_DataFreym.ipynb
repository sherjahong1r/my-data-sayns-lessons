# **Elementlarni tanlash DataFreym**

```python
import pandas as pd
import numpy as np
```

```python
uz = {2025:36.1, 2024:35.8, 2023:34.4, 2022:34.1, 2021:33.9, 2020:33.5}
kz = {2025:18.8, 2024:18.5, 2023:18.3, 2022:18.1, 2021:17.8, 2020:17.6}
tj = {2025:9.5, 2024:9.3, 2023:9.1, 2022:8.9, 2021:8.7, 2020:8.5}
kg = {2025:6.5, 2024:6.3, 2023:6.2, 2022:6.1, 2021:6.0, 2020:5.9}
tr = {2025:6.1, 2024:6.0, 2023:5.9, 2022:5.8, 2021:5.7, 2020:5.6}

df = pd.DataFrame(
    {"Uzbekistan":uz,
     "Kazakhstan":kz,
     "Tajikistan":tj,
     "Kyrgizstan":kg,
     "Turkmenistan":tr
     }
)
```
