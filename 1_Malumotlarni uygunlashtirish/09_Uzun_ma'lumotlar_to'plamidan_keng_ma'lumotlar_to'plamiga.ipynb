```python
data = pd.read_csv("https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/refs/heads/master/data_pivot.csv")
data
# Bu to'g'ridan to'g'ri git hub dan olib kelindi lekin ikkalasi ham bir xil.
```

```python
pivoted_data = data.pivot(index='date', columns='item', values='value')
pivoted_data
# yuqorida asl jadval 609 qator edi, ammo bundan so'ng 203 qator bo'ldi yani takrorlanganlar tashlab yuborildi
# infl	realgdp	unemp ustunlari paydo bo'lib qoldi bular yuqoridagi jadvalning item ustunidagi mavjud bo'lgan buyumlar
# bu yerda ularni alohida ustun qilganimiz uchun ham qatorlar soni kamaygan va item nomi to'g'risida turibdi
# values ustuni qiymatlari esa bu (infl	realgdp	unemp) ustunlar qatorlarida joylashgan.
```

```python
data['value_2'] = np.random.randn(len(data))
data
# Taxminiy qiymatlardan iborat data_2 degan ustun qo'shildi asl data jadvaliga
```

```python
pivoted_data = data.pivot(index='date', columns='item', values=['value', 'value_2'])
pivoted_data
# iyerarxik ko'rinishga keldi.
```
