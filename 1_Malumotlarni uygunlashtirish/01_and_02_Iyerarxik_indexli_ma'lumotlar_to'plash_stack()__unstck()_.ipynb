## **Hierarchical indexing - Iyerarxik indekslash**

```python
data = pd.Series(np.random.randn(9),index=['a','a','a','b','b','c','c','d','d']) # odatiy indeksli Series
data

# random bilan 9 ta son va uning indexsiga yuqoridagi harflar qo'yildi
```

```python
data = pd.Series(np.random.randn(9),
                 index=['a','a','a','b','b','c','c','d','d']) # odatiy indeksli Series
data.index
```

```python
data
# data ning indexlari chiqarildi
```

```python
data1 = pd.Series(np.random.randn(9),
                  index=[['a','a','a','b','b','c','c','d','d'],[1,2,3,1,3,1,2,2,3]])  # irarxik indeksli Series
data1

# Bu orqali guruhlab olyabmiz yani raqamlardan 1,2,3 - a guruhiga. 1,3 - b.  1,2 - c. 2,3 - d guruhlariga to'g'ri kelyabdi.
# Yani sonli indexlar irarxik indekslar bilan guruhlanayabdi
```

```python
data.index # Bunday holatga indeks deyiladi
```

```python
data1.index  # Bunga esa MultiIndex deyiladi
```

```python
data1
```

```python
data1['a']
# subset bilan ajratib olish
# yani a guruhidagi sonli indexlar qatorini chiqaradi
```

```python
data1['b':'d']
# Slicing yani kesib olyabdi b dan d gacha bo'lgan qismini ajratib olyabdi
```

```python
data1['a':'c']
# a dan c gacha qismi chop etildi
```

```python
data1.loc[:,2]
# indexi 2 bo'lgan qatorni chiqaradi
```

```python
data1.loc[['b','d']]
# b va d guruhidagi indexlar qatorini chiqaradi
```

```python
data1.loc[:,1]
# qaysi harfli guruhda 1 indexi bo'lsa shuni chiqaradi.
```

```python
df1 = data1.unstack()
df1
# Sochilgan Seriesni df ko'rinishida qaytaradi.
# Etibor bersak indekslar column yani ustun bo'lib qoldi.
# Hamda b, c, d dagi bitta kam bo'lgan qiymatlar o'rni NaN bo'lib turibdi
```
