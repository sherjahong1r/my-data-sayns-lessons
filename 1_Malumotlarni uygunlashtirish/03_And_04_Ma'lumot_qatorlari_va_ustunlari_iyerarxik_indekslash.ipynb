```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df
# Etibor bersak City dan tashqari qolgan ustunlar qo'shtirnoq ichida edi
# uni to'g'irlash uchun hammasiga qaytadan nom berdik
```

```python
# Yuqoridagi jadvalga etibor bersak Region ustunidagi qatorlar qo'shtirnoq ichiga yozilgan
df['Region'] = df['Region'].map(lambda x: x.replace('"', ''))
df
# Buni yo'q qilish uchun .map bilan yani u ma'lumotlarni tozalash, formatlash uchun  Pandas DataFrame ustunidagi
# (Series obyekti) har bir elementga biror funksiyani qo'llash uchun ishlatiladi.
# map ichida lambda bilan hamma qiymatlarni x qiladi va uni yani bu x qiymatlarni x.replace yani almashtirish degani
# u bilan bir qo'sh tirnoq ichidagi qo'shtirnoqli qatorlar degani va uni almashtiradi verguldan keyingi natijaga yani
# u bir tirnoq ichida hech narsa bo'lmagani uchun qo'shtirnoqlarni tashlab yuboradi qiymatlardan.
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"', ''))
df
# Endi Country ustunidan ham qo'sh tirnoqlarni olib tashladik
```

## pivot_table orqali irarxik indekslash

```python
df
```

```python
df.index
# Bular tartibli oddiy indexlar
```

```python
# Yuqoridagi jadvalga etibor bersak bir davlatlar ko'p takrorlangan uning shaharlari ko'p bo'lhanligi uchun
# endi bunda davlatlarni irarxik yani index guruhi qilib qo'yamiz pivot_table yordamida

df3 = pd.pivot_table(df, index=['Country', 'City'], values=['AirQuality', 'WaterPollution'], aggfunc='mean')
df3

# Country (davlat) asosiy indeks bo'lib, uning ichida City (shahar) ikkilamchi indeks bo'ladi. Bu iyerarxik indekslashni (multi-index) hosil qiladi,
# ya'ni har bir davlat ostida uning shaharlari guruhlanadi. values=['AirQuality', 'WaterPollution']: Bu parametr qaysi ustunlardagi
# ma'lumotlar jamlanishini (aggregate) ko'rsatadi. aggfunc='mean': Bu parametr values da ko'rsatilgan ustunlarga qanday jamlash (aggregation)
# funksiyasi qo'llanilishini aniqlaydi.  qiymatlarining o'rtacha (mean) miqdorini hisoblaydi.
```

```python
df3['AirPollution'] = 100-df3['AirQuality']
df3
# AirPollutionga saqlanayabdi 100 dan ayrilgan df3 AirQuality qiymatlari

# bu iyerarxik qilgan jadvalimiz bo'lib
# Country - mamlakatlar, city - shaharlar.
# AirQuality - havo sifati bo'lsa, AirPollution - havo ifloslanishi hisoblanadi va shu ustunni hozir yaratdik
# Misol uchun Afganistanning Herat shahrida havo sifati 43.75 bo'lsa, ifloslanishi 56.25 yani 100 % lik meyordan olingan
# va ikkalasining yig'indisi 100 bo'ladi.
# WaterPollution esa suv ifloslanishi.
```

```python
df3['WaterQuality'] = 100-df3['WaterPollution']
df3
# Endi huudi yana shunday WaterQuality jadval ustuni qo'shildi
```

```python
df3 = pd.DataFrame(df3.values, index = df3.index, columns= [['AIR','AIR','WATER','WATER'], ['AirQuality','AirPollution','WaterQuality','WaterPollution']])
df3
# Bunda ustunlarni iyerarxik ko'rinishga o'tkazdik
```
