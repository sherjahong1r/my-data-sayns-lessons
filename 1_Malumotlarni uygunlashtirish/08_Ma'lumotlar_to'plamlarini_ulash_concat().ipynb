```python
bank = pd.read_csv("/content/Practicum_Datasets/bank.csv")
bank
# bu tartibsiz holda joylashgan
```

```python
bank = pd.read_csv("/content/Practicum_Datasets/bank.csv", delimiter=';')
bank.head()
# delimiter=';' bilan tartibli holda chiqarayabmiz
```

```python
bank.info()
```

```python
bank1, bank2 = bank[:2521], bank[2521:]
# Bu bilan ikki qismga ajratib olyabmiz yani 2521 tagacha qatorni bank1 ga va qolgan 2000 qatorini bank2 ga saqladik
```

```python
bank1
```

```python
bank2
```

```python
pd.concat([bank1,bank2])
# 2 ta datasetni ulash yani yana qaytadan bir biriga qo'shib qo'ydik
```

```python
bank1, bank2, bank3, bank4 = bank[:1000], bank[1000:2000], bank[2000:3000], bank[3000:]
 # bunda 4 ta qisimga ajratib oldik
```

```python
bank1
```

```python
bank2
```

```python
bank3
```

```python
bank4
```

```python
pd.concat([bank1,bank2,bank3,bank4])
# concat bilan yana yuqorida ajratgan 4 datasetsni ulayabmiz
```

```python
pd.concat([bank1,bank2,bank3,bank4], axis=0)
# Bu ham yuqoridagi bilan bir hil yani qatorlarni tagma tag qilib ulaydi yuqoridagi barchasi ham.
```

```python
pd.concat([bank1,bank2,bank3,bank4], axis=1)
# axis=1 bilan esa gorizontaliga davom etadi yani ulanadi.
# Bu yerdagi NaN qiymatlar esa bank1 1000 gacha edi, bank2 esa 1000 dan 2000 gacha edi
# shunda bank2 bank1 dan pastki yonida va o'ng tomonida joylashadi va uning yuqori tepa qismi va chap yon tomoni
# qatorlari NaN bilan to'ldirilib ketiladi. Qolgan datasetlar ham huddi shu tartibda bo'ladi.
# Shu sababli NaN lar vujudga kelgan
```
