## Ko'paytirish, bo'lish, qo'shish, ayirish, va h.k amallar

```python
# data
arr1 = np.arange(6).reshape(2, 3) # 0~6 gacha bo'lgan elementlardan tashkil topgan array
arr2 = np.arange(6, 12).reshape(2, 3) # 6dan 11 gacha bo'lgan elementlardan tashkil topgan arrey
```

```python
arr1
```

```python
arr2
```

```python
# Ko'paytirish
arr1 * arr2 # arr1 ning mos elementlarini arr2 ning mos elementlariga ko'paytirib beradi
```

```python
# Bo'lish
arr2/arr1  # arr1 ning mos elementlarini arr2 ning mos elementlariga bo'lib beradi. Hamda 0 bo'lib bo'lmagani uchun inf bilan xato berayabdi
```

```python
# Qo'shish
arr1 + arr2 # arr1 ning mos elementlarini arr2 ning mos elementlariga qo'shib beradi
```

```python
# Ayirish
arr2-arr1 # arr2 ning mos elementlarini arr1 ning mos elementlaridan ayirib beradi
```

```python
# kvadratga oshirish (sqare)
arr1**2
```

```python
# kvadrat ildizdan (sqrt)
arr2**0.5
```
