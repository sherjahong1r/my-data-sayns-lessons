```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("new.db") # bu new.db MO ini db deb saqladik
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpopulation(
rank INT,
country TEXT PRIMARY KEY,
density DOUBLE,
population DOUBLE NOT NULL,
area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
query = "INSERT INTO worldpopulation VALUES(1, 'Uzbekistan', 33000, 33000000, 445000)"

cursor.execute(query)
db.commit()
```

```python
query_select = "SELECT * FROM worldpopulation"
df = pd.read_sql_query(query_select, db)
display(df)
```
