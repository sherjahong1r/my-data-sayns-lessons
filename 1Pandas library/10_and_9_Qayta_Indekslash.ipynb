```python
import pandas as pd
import numpy as np
```

```python
data = {"Yil":[2025, 2024, 2023, 2022, 2021, 2020, 2019,2014],
        "Aholi soni":[36.5, 35.4, 35.1, 34.3, 33.1, 32.4, 31.5, 30.8],
        "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        "Zichlik":[80.23, 79.45, 77.34, 75.13, 73.34, 71.23, 70.23, 68.23]}
df = pd.DataFrame(data)
df
```

```python
df.index = [25, 24, 23, 22, 21, 20, 19, 14]
df
```

```python
# reindex
ind = [25, 24, 23, 22, 21, 20, 19, 14]
ind.sort() # Bu bilan kichikdan kattaga o'sadi
ind
```

```python
df1 = df.reindex(ind)
df1
```

```python
df
```

```python
df2 = df.reindex([14,20,25]) # Biz tanlagan yillarni chiqaradi
df2
```

```python
df3 = df.reindex(index=range(14,25)) # range bilan 14 dan 25 yilgacha chiqaradi, hamda orasida yo'q qiymatlarni nan qilib ketadi.
df3
```

```python
df4 = df.reindex(index=range(14,25), method='nearest') # bu bilan nan qiymatlar o'rnini yuqoridagiga yaqin bo'lgan qiymatlar bilan to'ldiradi
df4
```

```python
df5 = df.reindex(index=range(14,25), method='ffill') # Bu bilan nan o'rnini faqat bitta pasdagi qiymat bilan to'ldiradi.
df5
```

```python
df1
```

```python
df6 = df.reindex(columns=['Aholi soni', 'Temp', 'Zichlik']) # Bu bilan biz yozgan ustunlarni chiqaradi
df6
```

```python
df7 = df6.head() # Boshidagi 5 tani chiqaradi
df7
```

```python
df8 = df6.sample(2) # Tasodifiy qiymatni olib beradi, Qiymatiga qarab
df8
```

```python
df9 = df.loc[[14,20,23,25], ['Aholi soni', 'Temp', 'Zichlik']] # indexlar bilan chiqarishning yana bir usuli
df9
```
