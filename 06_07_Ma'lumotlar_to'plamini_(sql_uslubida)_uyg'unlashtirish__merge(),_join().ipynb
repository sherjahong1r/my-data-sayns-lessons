## Sodda misollar

```python
A = pd.DataFrame({'Transport': ['poezd','avtomabil','kema',],
                  'Tezligi':['70','180','45']})
A
```

```python
B = pd.DataFrame({'Transport': ['poezd','avtomabil','tramvay'],
                  "Narxi":['98000','10000','65000']})
B
```

```python
merged = pd.merge( A, B)  # inner join - ichki birlashish degani
merged
# merge da har ikkala jadvalni birlashtirganimizda qo'shimcha argumentlarsiz yani buyruqlarsiz foydalanadigan bo'lsak
# faqat ikkalasida bir xil mavjud bo'lgan qiymatlarnigina chiqaradi holos
# lekin movies_ratings ni birlashtirganimizda bunday bo'lmagan edi sababi bularni bog'lab turuvchi umumiy bitta ustun va qatorlar
# sonigacha bir biriga muozanatlashgan edi.
```

```python
merged = pd.merge( A, B, how = 'inner')  # bu ham yuqoridagi natijani beradi va ichki birlashish hisoblanadi
merged
```

```python
merged_outer = pd.merge(A, B, how='outer') # outer join - tashqi
merged_outer
# Bunga qo'shimcha argument tariqasida how='outer' qo'shilgani uchun qolgan qatorlarni ham chiqarib berdi
# etibor bersak kema ning narxi degan ustuni bo'lmagani uchun uning o'rniga NaN qaytaradi qolganlari ham shunday tartibda.
```

```python
A
```

```python
B
```

```python
merged_left = pd.merge(A, B, how='left')  # left join - chap birlashish
merged_left
# bunda how='left' bilan dastlab ikkalasida bor bo'lgan umumiy xususiyatlar olinadi so'ng faqat chap tomonda joylashgan
# A jadvalining mavjud qiymatlari olinadi.
# umumiylari poezd va avtomabil bo'lsa A jadvalidagi ortiqcha qiymat kema edi va bu qo'shildi.
```

```python
merged_right = pd.merge(A, B, how='right')  # right join - o'ng birlashish
merged_right
# Bunda esa umumiylardan so'ng o'ng tarafdagi B ning qatorlari chiqariladi
```

## merge bilan ikkala jadvalda ham bir xil umumiy jadvallar bo'lgani uchun ham hatosiz chiqmoqda.  
## Agar umumiy ustunlar bo'lmaganda nima bo'lardi buni A jadvalini o'zgartirib tekshirib ko'ramiz

```python
A = pd.DataFrame({'Transport_turi': ['poezd','avtomabil','kema',],
                  'Tezligi':['70','180','45']})
A
```

```python
B
```

```python
# pd.merge(A. B)

## Bunday qilsak xatolik yuzaga keladi chunki umumiy birlashtiruvchi ustunlar mavjud emas.
```

```python
merge = pd.merge(A, B, left_on='Transport_turi', right_on='Transport')
merge
# Bunda chapdagi A jadvalidan left_on bilan Transport_turi ustuni va o'ngdagi B jadvalidan right_on bilan
# Transport ustunini birlashtirish kerak ekanligini aytyabmiz
# lekin bunda ikkta ustunni ham chiqarganligi bois bittasini o'chirish kerak bo'ladi
```

```python
merge = merge.drop('Transport', axis=1)
merge
```
